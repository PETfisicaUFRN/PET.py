## 1. Introdu√ß√£o ao efeito Compton

### 1.1 O que √© o efeito Compton?

$\quad$ Sabemos que um el√©tron pode absorver energia de um f√≥ton, mas o el√©tron tamb√©m pode perder energia emitindo na forma de f√≥tons. Um el√©tron imerso em um campo el√©trico acelerado por ele pode produzir um ou v√°rios f√≥tons (quando colidir com a mat√©ria). Ent√£o, caso a diferen√ßa de potencial aceleradora obtenha valores nas casas das dezenas ou centenas de kV, o f√≥ton ou os f√≥tons que s√£o gerados v√£o ter uma energia de mesma faixa de intensidade. Para esse tipo de radia√ß√£o eletromagn√©tica, demos o nome de raio X. 

$\quad$ Os raios X tem esse nome devido ao seu descobridor Roentgen n√£o conhecer a natureza dessa onda eletromagn√©tica, que possui um comprimento de onda menor do que 1√Ö. Como qualquer onda, os raios X, apresentam propriedades que s√£o caracter√≠sticas das ondas, como a interfer√™ncia, a difra√ß√£o e a polariza√ß√£o. Cargas el√©tricas, que est√£o sob a a√ß√£o de um campo eletromagn√©tico incidente, ganham um movimento de oscila√ß√£o de frequ√™ncia que √© o mesmo do campo. Deste modo, produzem e emitem uma radia√ß√£o com uma frequ√™ncia igual ao seu movimento oscilat√≥rio.

$\quad$ Agora, em um ponto de vista corpuscular (tratado como part√≠cula), o espalhamento de f√≥tons seria visto como sendo a colis√£o de um f√≥ton de um raio X com uma part√≠cula. Deste modo, acarretaria na transfer√™ncia de momento do f√≥ton incidente para a part√≠cula em quest√£o. Uma part√≠cula em repouso, por exemplo, um el√©tron, ao colidir com o f√≥ton, iria adquirir momento e consequentemente energia cin√©tica. Logo, diminuindo a energia do f√≥ton. Na quest√£o ondulat√≥ria, o que acontece √© diminui√ß√£o da frequ√™ncia e aumento do comprimento de onda.

$\quad$ Em resumo, o efeito Compton se trata do espalhamento de um f√≥ton devido a um choque com uma part√≠cula carregada, que geralmente se trata do el√©tron, que por sua vez resulta na diminui√ß√£o de energia devido ao aumento do comprimento de onda do f√≥ton espalhado.

Este, por sua vez, √© chamado de deslocamento Compton,¬†ŒîùúÜ=ùúÜ‚Ä≤‚àíùúÜ¬†e se altera conforme o √¢ngulo no qual os raios X s√£o visualizados.

Do ponto de vista cl√°ssico, o campo el√©trico oscilante com uma certa frequ√™ncia¬†ùë£¬†da onda incidente age sobre os el√©trons livres, fazendo com que eles oscilem com a mesma frequ√™ncia.

Deste modo, Ele observou que os raios X espalhados apresentavam um deslocamento de comprimento de onda, chamado de efeito Compton, que n√£o poderia ser explicado pela teoria cl√°ssica das ondas eletromagn√©ticas. Esse desvio s√≥ p√¥de ser compreendido pela teoria qu√¢ntica da radia√ß√£o, que descreve a luz como uma corrente de part√≠culas, os f√≥tons. O experimento de Compton forneceu evid√™ncias cruciais para a dualidade onda-part√≠cula da luz e teve aplica√ß√µes significativas em √°reas como medicina e f√≠sica de altas energias.

## 2. Teoria do Efeito Compton

$\quad$ Analisando os dados experimentais, Compton postulou que os feixes de raios X n√£o se tratavam de uma onda de frequ√™ncia¬†ùë£, e sim, um conjunto de f√≥tons, com energia dada por¬†ùê∏=‚Ñéùë£, e s√£o esses f√≥tons que colidem com os el√©trons livres do alvo. Ent√£o, a radia√ß√£o que √© espalhada se trata de f√≥tons que colidiram com esses el√©trons livres. Deste modo, visto que o el√©tron absorve parte da energia do f√≥ton incidente, o f√≥ton espalhado deve ter uma energia¬†ùê∏‚Ä≤¬†menor, logo, ele deve obter uma frequ√™ncia mais baixa¬†ùë£‚Ä≤=ùê∏‚Ä≤/‚Ñé, o que, por sua vez, implica num comprimento de onda¬†ùúÜ‚Ä≤=ùëê/ùë£‚Ä≤¬†maior. Note que, na intera√ß√£o de f√≥ton-el√©tron, s√£o vistos como part√≠culas, e n√£o como ondas, e que, diferentemente do efeito fotoel√©trico, os f√≥tons s√£o espalhados e n√£o absorvidos.

$$E = hf$$

Sendo:

* E - Energia do f√≥ton
* h - Constante de Planck
* $f$ - Frequ√™ncia do f√≥ton

Considerando a ideia de que o f√≥ton √©, na verdade, um pacote localizado de energia, iremos compreend√™-lo como uma part√≠cula de energia¬†ùê∏¬†e momento¬†ùëù. Agora, considerando a energia relativ√≠stica dada pela equa√ß√£o:

$$E = \frac{m_0c^2}{\sqrt{1-\frac{v^2}{c^2}}}$$

Visto que o f√≥ton viaja a uma velocidade c (velocidade da luz), e tem E = hf, implica que ele n√£o tem massa, portanto sua energia de repouso √© igual a zero. Sendo assim, √© poss√≠vel considerar o f√≥ton como sendo uma part√≠cula de massa nula e de energia relativ√≠stica total¬†ùê∏, sendo est√° energia totalmente cin√©tica.

O momento do f√≥ton pode ser obtido a partir da rela√ß√£o de energia total relativ√≠stica E, momento p e massa de repouso¬†$m_0$.

A rela√ß√£o √© dada por:

$$E^2 = c^2p^2 + (m_0c^2)^2 \tag{1.1}$$

O termo $m_0c^2$ √© nulo quando se trata do f√≥ton, uma vez que a massa dele √© nula. Logo:

$$p = \frac{E}{c} = \frac{hf}{c} $$

ent√£o:

$$p = \frac{h}{\lambda}$$

$\lambda = \frac{c}{\lambda}$ se trata do comprimento de onda. 

### 2.1 Equa√ß√£o do deslocamento Compton

<center>
    
<center>

$\quad$ Considerando uma colis√£o entre um f√≥ton e um el√©tron livre em repouso. Podemos encontrar a express√£o para o efeito Compton. Sabendo que o f√≥ton viaja na velocidade da luz, √© mais que evidente que estamos lidando com limites n√£o relativ√≠sticos. Logo, o f√≥ton possui uma energia total relativ√≠stica¬†$E_0$¬†e momento¬†$p_0$¬†que incide em um el√©tron em repouso que possui uma energia de repouso de¬†$m_0c^2$. Analisando a imagem acima, vemos que o f√≥ton √© espelhado em um √¢ngulo¬†ùúÉ¬†e se afasta com uma energia¬†$E_1$¬†e momento¬†$p_1$, enquanto o el√©tron recua, formando um √¢ngulo¬†ùúô¬†com o eixo da colis√£o, com energia cin√©tica $K$ e momento $p$. Aplicando a convers√£o de momento e de energia, temos que:

Conserva√ß√£o de momento:

No eixo x

$$p_0 = p_1cos(\theta)+p \hspace{0.1cm} cos(\phi)$$

No eixo y

$$p_1 sin(\theta) = p \hspace{0.1cm} sin(\phi)$$

Elevando as equa√ß√µes ao quadrado, temos que

$$(p_0 - p_1 \hspace{0.1cm}cos(\theta))^2 = p^2 \hspace{0.1cm}cos^2(\phi)$$

e

$$p_1^2 sin^2(\theta) = p^2 \hspace{0.1cm} sin^2(\phi)$$

Somando as duas equa√ß√µes, encontramos que

$$p_0^2 + p_1^2 - 2p_0p_1\hspace{0.1cm}cos(\theta) = p^2 \tag{2.1}$$

Utilizando a conserva√ß√£o da energia total relativ√≠stica, temos que

$$E_0 + m_0c^2 = E_1 + K +m_0c^2$$

Visto que as energias de repouso antes da colis√£o e depois da colis√£o s√£o iguais, temos que

$$E_0 - E_1 = K $$

Isso pode ser reescrito da seguinte forma

$$c(p_0 - p_1) = K \tag{2.2}$$

Usando $E = K + m_0c^2$ em (1.1), encontramos

$$(K + m_0c^2)^2 = c^2p^2 + (m_0c^2)$$

Simplificando

$$K^2 + 2Km_0c^2 = p^2$$

Substituindo¬†$p^2$¬†pela equa√ß√£o (2.1) e K pela equa√ß√£o (2.2), na equa√ß√£o que encontramos acima, temos que

$$(p_0 - p_1)^2 + 2m_0c(p_0 - p_1) = p_0^2 + p_1^2 - 2p_0p_1 \hspace{0.1cm}cos(\theta)$$

Expandindo $(p_0 - p_1)^2$ obtemos

$$(p_0 - p_1)^2 = p_0^2 - 2p_0p_1 + p_1^2$$

Deste modo, podemos reduzir a equa√ß√£o para

$$m_0c(p_0 - p_1) = p_0p_1(1-cos(\theta))$$

ou 

$$\frac{1}{p_1} - \frac{1}{p_0} = \frac{1}{m_0c}(1-cos(\theta)) \tag{2.3}$$

Multiplicando a equa√ß√£o acima por h temos que

$$\Delta \lambda = \lambda_1 - \lambda_0 = \frac{h}{m_0c}(1 - cos(\theta)) \tag{2.4}$$

O termo $\frac{h}{m_0c}$ tem como resultado um valor de $2,43 \times 10^{-12}m = 0,0243A = 2,43pm$. Este valor, por sua vez √© chamado de comprimento de Compton. Obtida a equa√ß√£o de Compton √© poss√≠vel observar que o deslocamento de Compton depende apenas do espalhamento $\theta$,e n√£o do comprimento de onda inicial $\lambda_0$.

## 3.1 Importando as bibliotecas necess√°rias

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import Image
from scipy.constants import Planck,speed_of_light, electron_mass
```

## 3.2 vers√£o dos softwares

Nesse c√≥digo, vamos observar o aumento do comprimento de onda em fun√ß√£o do angulo theta.

*Constantes*

```python
c = speed_of_light 
h = Planck
m = electron_mass
```

Utilizando a equa√ß√£o de compton que deduzimos acima, temos que:

```python
theta = np.arange(0.,1.,1/180)*np.pi # Angulo theta em pi-radianos
delta_lambdaa = 2.43e-12*(1-np.cos(theta)) #equa√ß√£o de Compton
theta_graus = np.arange(0,180) # Angulo theta em graus para plotar no gr√°fico
print(delta_lambdaa)
```

Deste modo, √© not√≥rio que o deslocamento Compton varia conforme o √¢ngulo theta varia.

```python
# comprimento de onda do f√≥ton incidente
plt.figure(figsize=(9,6))
plt.style.use('dark_background')
plt.grid('--')
plt.plot(theta_graus,delta_lambdaa,color='red')
plt.title('Gr√°fico deslocamento Compton em fun√ß√£o do angulo de espalhamento')
plt.xlabel('√Çngulos (¬∫)')
plt.ylabel([0,6e-12])
plt.ylabel('Deslocamento Compton')
```

$\quad$ O gr√°fico apresentado acima mostra a rela√ß√£o entre o deslocamento Compton e o √¢ngulo theta. O efeito Compton √© uma consequ√™ncia direta da natureza dual das part√≠culas e das ondas. Este gr√°fico oferece uma visualiza√ß√£o clara da depend√™ncia do deslocamento Compton em rela√ß√£o ao √¢ngulo de espalhamento.

Podemos tamb√©m deixar o gr√°fico um pouco mais din√¢mico fazendo uma anima√ß√£o.

```python
c = speed_of_light 
h = Planck
lambdaa_RaioX = 10e-9
m = electron_mass
theta_graus = np.arange(0,180)

fig, ax = plt.subplots()

x_dados, y_dados = [],[]
ln, = plt.plot([],[],'r-')

def init():
    ax.set_xlim([0,180])
    ax.set_ylim([0,5.5e-12])
    return ln,

def atualizar(frame):
    theta_rad = np.radians(frame)
    lambdaa_linha = 2.43e-12*(1-np.cos(theta_rad)) #equa√ß√£o de Compton
    y_dados.append(lambdaa_linha)
    x_dados.append(frame)
    ln.set_data(x_dados,y_dados)
    return ln,

ani = animation.FuncAnimation(fig,atualizar,frames=theta_graus,init_func=init,blit=True)
ani.save('compton.gif',writer='imagemagick',fps=30)
Image(filename='compton.gif')
```

## Extra

$\quad$ Antes do f√≥ton se chocar com a part√≠cula, vimos que a energia do f√≥ton √© dada por¬†ùê∏=‚Ñéùëì. No espalhamento Compton, a energia do f√≥ton ap√≥s a colis√£o pode ser calculada pela equa√ß√£o abaixo:

$$E_1 = \frac{E_0}{(\frac{E_0}{mc^2})(1 - cos\theta)+1}$$

Podemos deduzi-l√° por conserva√ß√£o de momento e energia, juntamente com princ√≠pios da relatividade.

Multiplicando a equa√ß√£o (2.1) por $\frac{1}{c}$, temos que:

$$\frac{1}{cp_1} - \frac{1}{cp_0} = \frac{1}{m_0c^2}(1-cos(\theta))$$ 

Visto que $cp_1$ se trata da energia ap√≥s a colis√£o e $cp_0$ se trata da energia antes da colis√£o, temos que

$$\frac{1}{E_1} - \frac{1}{E_0} = \frac{1}{m_0c^2}(1-cos(\theta))$$ 

Desenvolvendo a equa√ß√£o

$$\frac{1}{E_1} = \frac{1}{m_0c^2}(1-cos(\theta)) + \frac{1}{E_0}$$ 

$$ 1 = \left[\frac{1}{m_0c^2}(1-cos(\theta)) + \frac{1}{E_0}\right]E_1$$

$$ 1 = \left[\frac{E_1(1-cos(\theta)) + mc^2}{m_0c^2E_0}\right]E_1 $$

Agora, isolando $E_2$ e simplificando a equa√ß√£o chegamos em:

$$E_1 = \frac{E_0}{(\frac{E_1}{mc^2})(1 - cos\theta)+1}$$

$\quad$ No c√≥digo abaixo, definimos a energia inicial do f√≥ton como E1 e calculamos a energia do f√≥ton ap√≥s a colis√£o E2 para cada √¢ngulo de espalhamento. Em seguida, colocamos a rela√ß√£o entre o √¢ngulo de espalhamento e a energia do f√≥ton ap√≥s a colis√£o. Isso nos permitir√° visualizar como a energia do f√≥ton varia com o √¢ngulo de espalhamento, conforme a equa√ß√£o do Efeito Compton.

```python
f = 10**20 # frequ√™ncia do raio x em Hz
E1 = h*f
E2 = E1 / (1 + (E1 / (m * c**2)) * (1 - np.cos(theta)))

theta_graus = np.degrees(theta)
```

```python
plt.plot(theta_graus,E2,color='blue')
plt.grid(True)
plt.xlabel('√Çngulo de Espalhamento (graus)')
plt.ylabel('Energia do F√≥ton ap√≥s a Colis√£o (J)')
plt.title('Rela√ß√£o entre √Çngulo de Espalhamento e Energia do F√≥ton')
plt.show()
```
