## 1. Introdução

### O que é o efeito Compton?

$\quad$ Sabemos que um eletrón pode ganhar/absorver energia de um fotón, mas o eléton também pode perder energia emitindo na forma de fotóns. Um eletro imerso em um campo elétrico que é acelerado por ele pode produzir um ou vários fotóns(quando colidir com a matéria). Então, caso a diferença de potencial aceleradora obtenha valores nas casas das dezenas ou centenas de kV, o fotón ou os fotóns que são gerados vão ter uma energia de mesma faixa de intensidade. Para esse tipo de radiação eletromagnética demos o nome de raio X. 

$\quad$ Os raios X tem esse nome devido ao seu descobridor Roentgen não conhecer a natureza dessa onda eletromagnetica, que possiu um comprimento de onda menor do que 1Å. Como qualquer onda, os raios X, apresentam propriedades que são caracterísicas da ondas como a interferência, a difração e a polarização. Cargas elétricas, que estão sob a ação de um campo elétromagnetico incidente, ganham um movimento de oscilação de frequência que é o mesmo do campo. Deste modo, produzem e emitem uma radiação com uma frequência igual ao seu movimento oscilatorio.

$\quad$ Agora, em um ponto de vista corpuscular(tratado como partícula), o espalhamento de fótons seria visto como sendo a colisão de um fóton de um raio X com uma partícula. Deste modo, acarretaria na transferência de momento do fóton incidente para a partícula em questão. Um partícula em repouso, por exemplo um eletron, ao colidir com o fóton iria adquirir momenton e consequentemente adquiri energia cinética. Logo, diminuindo a energia do fóton. Na questão ondulatória, o que acontece é diminuição da frequência e aumento do comprimento de onda.

$\quad$ Em resumo, o efeito compton se trata do espalhamento de um fotón devido a um choque com uma partícula carregada, que geralmente se trata do elétron, que por sua vez, resulta na diminuição de energia devido ao aumento do comprimento de onda do fotón espalhado.

## 1.2 História da descoberta do Efeito Compton

$\quad$ O físico Arthur Holly Compton(1892-1962), confirmou em 1923, a natureza corposcular da radiação. Ele utilizou um feixe de raios X de comrpimento de onda $\lambda$ e incidiu o feixe sobre um alvo de grafite, como o mostrado na figura abaixo. Ele mediu a intensidade dos raios X que foram espalhados em função do seu comprimento de onda para vários angulos de espalhamento.


<center>
    
<center>

$\quad$ Ele analisou que o feixe incidente consistia de um unico comrpimento de onda $\lambda$, os raios X que foram espalhados tem maximos de intensidade em dois comrpimentos de onda, um comprimento de onda tem o mesmo comrpimento de onda do raio incidente e o outro $\lambda'$ é um comprimento de onda maior que o $\lambda$ por uma quantidade de $\Delta \lambda$. Este, por sua vez é chamado de, deslocamento Compton, $\Delta \lambda = \lambda' - \lambda$ e se altera de acordo com angulo no qual os raios X são vizualizados.

A existencia do comprimento de onda $\lambda'$ não deveria ser possível se os raios X incidentes forem entendidos como sendo uma onda eletromagnética clássica. Do ponto de vista clássico o campo elétrico oscilante com uma certa frequência $v$ da onda incidente age sobre os elétrons livres fazendo com que eles oscilem com a mesma frequência. 


## 2. Teoria do Efeito Compton

$\quad$ Analisando os dados experimentais Compton postulou que os feixe de raios X não se tratava de uma onda de frequência $v$, e sim, um conjunto de fótons, com energia dada por $E = hv$, e são esses fótons que colidem com os elétrons livres do alvo. Deste modo, a radiação que é espalhada se trata de fótons que colidiram com esses elétrons livres. Deste modo, visto que o eletron absorve parte da energia do fóton incidente, o fóton espalhado deve ter uma energia $E'$ menor, logo, ele deve obter uma frequência mais baixa $v'= E'/h$, o que por sua vez, implica num comprimento de onda $\lambda' = c/v'$ maior. Note que, na interação de fótons são vistos como partículas, e não como ondas, e que diferentemente do efeito fotoelétrico, os fótons são espalhados e não absorvidos. 

A energia de um fóton é dado por:

$$E = hv$$

Sendo:

* E - Energia do fóton
* h - Constante de Planck
* $v$ - Frequência do fóton

Considerando a ideia de que o fóton é na verdade um pacote de energia, considerando como uma particula $E$ e momento $p$. Agora, considerando a energia relativistica dada pela equação:

$$E = \frac{m_0c^2}{\sqrt{1-\frac{v^2}{c^2}}}$$

$\quad$ Visto que o fóton viaja a uma velocidade c(velocidade da luz), e tem E = hv, implica que ele não tem massa, portanto sua energia de repouso é igual a zero. Sendo assim, é possível considerar o fóton como sendo como uma partícula de massa nula e de energia relativística total $E$ é totalmente energia cinética.  

O momento do fóton pode ser obtido a partir da relação de energia total relativistica E, momento p e massa de repouso $m_0$. 

A relação é dado por:

$$E^2 = c^2p^2 + (m_0c^2)^2$$

O termo $m_0c^2$ é nulo quando se trata do fóton, uma vez que a massa dele é nula. Logo:

$$p = \frac{E}{c} = \frac{hv}{c} $$

então:

$$p = \frac{h}{\lambda}$$

$\lambda = \frac{c}{\lambda}$ se trata do comprimento de onda. 

## Equação do deslocamento de Compton

<center>
    
<center>

$\quad$ Considerando uma colisão entre um fóton e um elétron livre em repouso. Podemos encontrar a espressão para o efeito Compton. Sabendo que o fóton viaja na velocidade da luz é mais que evidente que estamos lidando com limites não relativísticos. Logo, o fóton possui uma energia total relativística $E_0$ e momento $p_0$ que incide em um elétron em repouso que possui uma energia de repouso de $m_0c^2$. Analisando a imagem abaixo vemos que o fóton é espelhado em um angulo $\theta$ e se afasta com uma energia $E_1$ e momento $p_1$, enquanto que o elétron recua, formando um angulo $\phi$ com o eixo da colisão, com energia cinética K e momento p. Aplicando a consevração de momento e de energia, temos que:

Conservação de momento:

No eixo x

$$p_0 = p_1cos(\theta)+p \hspace{0.1cm} cos(\phi)$$

No eixo y

$$p_1 sin(\theta) = p \hspace{0.1cm} sin(\phi)$$

Elevando as equações ao quadrado, temos que

$$(p_0 - p_1 \hspace{0.1cm}cos(\theta))^2 = p^2 \hspace{0.1cm}cos^2(\phi)$$

e

$$p_1^2 sin^2(\theta) = p^2 \hspace{0.1cm} sin^2(\phi)$$

Somando as duas equações encontramos que

$$p_0^2 + p_1^2 - 2p_0p_1\hspace{0.1cm}cos(\theta) = p^2$$

Utilizando a conservação da energia total relativística temos que

$$E_0 + m_0c^2 = E_1 + K +m_0c^2$$

Visto que as energia de repouso antes da colisão e depois da colisão são iguais, temos que

$$E_0 - E_1 = K$$

Isso pode ser reescrito da seguinte forma

$$c(p_0 - p_1) = K$$

Usando $E = K + m_0c^2$, enocntramos

$$(K + m_0c^2)^2 = c^2p^2 + (m_0c^2)$$

Simplificando

$$K^2 + 2Km_0c^2 = p^2$$

Substituindo $p^2$ pela equação(x,x) e K pela equação (x,x), na equação que encontramos acimas temos que

$$(p_0 - p_1)^2 + 2m_0c(p_0 - p_1) = p_0^2 + p_1^2 - 2p_0p_1 \hspace{0.1cm}cos(\theta)$$

Expandindo $(p_0 - p_1)^2$ obtemos

$$(p_0 - p_1)^2 = p_0^2 - 2p_0p_1 + p_1^2$$

Deste modo, podemos reduzir a equação para

$$m_0c(p_0 - p_1) = p_0p_1(1-cos(\theta))$$

ou 

$$\frac{1}{p_1} - \frac{1}{p_0} = \frac{1}{m_0c}(1-cos(\theta))$$

Multiplicando a equação acima por h temos que

$$\Delta \lambda = \lambda_1 - \lambda_0 = \frac{h}{m_0c}(1 - cos(\theta))$$

O termo $\frac{h}{m_0c}$ tem como resultado um valor de $2,43 \times 10^{-12}m = 0,0243A = 2,43pm$. Este valor, por sua vez é chamado de comprimento de Compton. Obtida a equação de Compton é possível observar que o deslocamento de Compton depende apenas do espalhamento $\theta$,e não do comprimento de onda inicial $\lambda_0$.

# Código

$$E_2 = \frac{E_1}{(\frac{E_1}{mc^2})(1 - cos\phi)+1}$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.constants import Planck,speed_of_light, electron_mass

c = speed_of_light 
h = Planck
lambdaa_RaioX = 10e-9
theta = np.arange(0.1,1.,1/180)*np.pi
E1 = h*lambdaa_RaioX
m = electron_mass

E2 = E1 / (( E1 / (m * c**2) )*(1 - np.cos(theta) + 1))
theta_graus = np.degrees(theta)
```

```python
plt.plot(theta_graus,E2,color='blue')
plt.grid(True)
plt.show()
```

```python
# comprimento de onda do fóton incidente

lambdaa_linha = 2.43e-12*(1-np.cos(theta))-lambdaa_RaioX
theta_graus = np.arange(0,180)

fig, ax = plt.subplots()
#plt.figure(figsize=(9,8))
plt.style.use('dark_background')
plt.grid('--')
ax.plot(theta_graus,lambdaa_linha,color='red')
ax.set_title('Gráfico comprimento de onda em função do angulo de espalhamento')
ax.set_xlabel('Ângulos (º)')
ax.set_ylabel('Comprimento de onda')
```
