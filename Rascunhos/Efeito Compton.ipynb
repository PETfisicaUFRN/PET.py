## 1. Introdução ao efeito Compton

### 1.1 O que é o efeito Compton?

$\quad$ Sabemos que um elétron pode absorver energia de um fóton, mas o elétron também pode perder energia emitindo na forma de fótons. Um elétron imerso em um campo elétrico acelerado por ele pode produzir um ou vários fótons (quando colidir com a matéria). Então, caso a diferença de potencial aceleradora obtenha valores nas casas das dezenas ou centenas de kV, o fóton ou os fótons que são gerados vão ter uma energia de mesma faixa de intensidade. Para esse tipo de radiação eletromagnética, demos o nome de raio X. 

$\quad$ Os raios X tem esse nome devido ao seu descobridor Roentgen não conhecer a natureza dessa onda eletromagnética, que possui um comprimento de onda menor do que 1Å. Como qualquer onda, os raios X, apresentam propriedades que são características das ondas, como a interferência, a difração e a polarização. Cargas elétricas, que estão sob a ação de um campo eletromagnético incidente, ganham um movimento de oscilação de frequência que é o mesmo do campo. Deste modo, produzem e emitem uma radiação com uma frequência igual ao seu movimento oscilatório.

$\quad$ Agora, em um ponto de vista corpuscular (tratado como partícula), o espalhamento de fótons seria visto como sendo a colisão de um fóton de um raio X com uma partícula. Deste modo, acarretaria na transferência de momento do fóton incidente para a partícula em questão. Uma partícula em repouso, por exemplo, um elétron, ao colidir com o fóton, iria adquirir momento e consequentemente energia cinética. Logo, diminuindo a energia do fóton. Na questão ondulatória, o que acontece é diminuição da frequência e aumento do comprimento de onda.

$\quad$ Em resumo, o efeito Compton se trata do espalhamento de um fóton devido a um choque com uma partícula carregada, que geralmente se trata do elétron, que por sua vez resulta na diminuição de energia devido ao aumento do comprimento de onda do fóton espalhado.

## 1.2 Breve história da descoberta do Efeito Compton

$\quad$ O físico Arthur Holly Compton(1892–1962) confirmou, em 1923, a natureza corpuscular da radiação. Ele utilizou um feixe de raios-x de comprimento de onda 𝜆 e incidiu o feixe sobre um alvo de grafite, como o mostrado na figura abaixo. Ele mediu a intensidade dos raios-x que foram espalhados em função do seu comprimento de onda para vários ângulos de espalhamento.

$\quad$ Ele analisou que o feixe incidente consistia em um único comprimento de onda, 𝜆. Os raios-x que foram espalhados tinham máximos de intensidade em dois comprimentos de onda, um comprimento de onda tinha o mesmo comprimento de onda do raio incidente e o outro 𝜆′ é um comprimento de onda maior que o 𝜆 por uma quantidade de Δ𝜆. Este, por sua vez, é chamado de deslocamento Compton, Δ𝜆=𝜆′−𝜆 e se altera conforme o ângulo no qual os raios-x são visualizados.

$\quad$ A existência do comprimento de onda 𝜆′ não deveria ser possível se os raios-x incidentes forem entendidos como sendo uma onda eletromagnética clássica. Do ponto de vista clássico, o campo elétrico oscilante com uma certa frequência 𝑣 da onda incidente age sobre os elétrons livres, fazendo com que eles oscilem com a mesma frequência.

$\quad$ Deste modo, ele observou que os raios-x espalhados apresentavam um deslocamento de comprimento de onda, chamado de efeito Compton, que não poderia ser explicado pela teoria clássica das ondas eletromagnéticas. Esse desvio só pôde ser compreendido pela teoria quântica da radiação, que descreve a luz como uma corrente de partículas, os fótons. O experimento de Compton forneceu evidências cruciais para a dualidade onda-partícula da luz e teve aplicações significativas em áreas como medicina e física de altas energias.

## 2. Teoria do Efeito Compton

$\quad$ Analisando os dados experimentais, Compton postulou que os feixes de raios-x não se tratavam de uma onda de frequência 𝑣, e sim, um conjunto de fótons, com energia dada por 𝐸=ℎ𝑣, e são esses fótons que colidem com os elétrons livres do alvo. Então, a radiação espalhada se trata de fótons que colidiram com esses elétrons livres. Deste modo, visto que o elétron absorve parte da energia do fóton incidente, o fóton espalhado deve ter uma energia 𝐸′ menor, logo, ele deve obter uma frequência mais baixa 𝑣′=𝐸′/ℎ, o que, por sua vez, implica num comprimento de onda 𝜆′=𝑐/𝑣′ maior. Note que, na interação de fóton-elétron, são vistos como partículas, e não como ondas, e que, diferentemente do efeito fotoelétrico, os fótons são espalhados e não absorvidos.

$$E = hf \ .$$

Sendo:

* E - Energia do fóton;
* h - Constante de Planck;
* $f$ - Frequência do fóton.

$\quad$ Considerando a ideia de que o fóton é, na verdade, um pacote localizado de energia, iremos compreendê-lo como uma partícula de energia 𝐸 e momento 𝑝. Agora, considerando a energia relativística dada pela equação:

$$E = \frac{m_0c^2}{\sqrt{1-\frac{v^2}{c^2}}} \ .$$

$\quad$ Visto que o fóton viaja a uma velocidade c (velocidade da luz), e tem E = hf, implica que ele não tem massa, portanto sua energia de repouso é igual a zero. Sendo assim, é possível considerar o fóton como sendo uma partícula de massa nula e de energia relativística total 𝐸, sendo está energia totalmente cinética.

$\quad$ O momento do fóton pode ser obtido a partir da relação de energia total relativística E, momento p e massa de repouso $m_0$.

$\quad$ A relação é dada por:

$$E^2 = c^2p^2 + (m_0c^2)^2 \ . \tag{1.1}$$

$\quad$ O termo $m_0c^2$ é nulo quando se trata do fóton, uma vez que a massa dele é nula. Logo:

$$p = \frac{E}{c} = \frac{hf}{c} \ , $$

então:

$$p = \frac{h}{\lambda} \ ,$$

na qual, $\lambda = \frac{c}{\lambda}$ se trata do comprimento de onda. 

### 2.1 Equação do deslocamento Compton

$\quad$ Considerando uma colisão entre um fóton e um elétron livre em repouso. Podemos encontrar a expressão para o efeito Compton. Sabendo que o fóton viaja na velocidade da luz, é mais que evidente que estamos lidando com limites não relativísticos. Logo, o fóton possui uma energia total relativística $E_0$ e momento $p_0$ que incide em um elétron em repouso que possui uma energia de repouso de $m_0c^2$. Analisando a imagem acima, vemos que o fóton é espelhado em um ângulo 𝜃 e se afasta com uma energia $E_1$ e momento $p_1$, enquanto o elétron recua, formando um ângulo 𝜙 com o eixo da colisão, com energia cinética $K$ e momento $p$. Aplicando a conversão de momento e de energia, temos que:

$\quad$ Conservação de momento:

$\quad$ No eixo x

$$p_0 = p_1cos(\theta)+p \hspace{0.1cm} cos(\phi) \ .$$

$\quad$ No eixo y

$$p_1 sin(\theta) = p \hspace{0.1cm} sin(\phi) \ .$$

$\quad$ Elevando as equações ao quadrado, temos que

$$(p_0 - p_1 \hspace{0.1cm}cos(\theta))^2 = p^2 \hspace{0.1cm}cos^2(\phi)$$

e

$$p_1^2 sin^2(\theta) = p^2 \hspace{0.1cm} sin^2(\phi)$$

$\quad$ Somando as duas equações, encontramos que

$$p_0^2 + p_1^2 - 2p_0p_1\hspace{0.1cm}cos(\theta) = p^2 \ . \tag{2.1}$$

$\quad$ Utilizando a conservação da energia total relativística, temos que

$$E_0 + m_0c^2 = E_1 + K +m_0c^2 \ .$$

$\quad$ Visto que as energias de repouso antes da colisão e depois da colisão são iguais, temos que

$$E_0 - E_1 = K \ . $$

$\quad$ Isso pode ser reescrito da seguinte forma

$$c(p_0 - p_1) = K \ . \tag{2.2}$$

$\quad$ Usando $E = K + m_0c^2$ em (1.1), encontramos

$$(K + m_0c^2)^2 = c^2p^2 + (m_0c^2) \ . $$

$\quad$ Simplificando

$$K^2 + 2Km_0c^2 = p^2 \ .$$

$\quad$ Substituindo $p^2$ pela Equação **(2.1)** e K pela Equação **(2.2)**, na equação que encontramos acima, temos que

$$(p_0 - p_1)^2 + 2m_0c(p_0 - p_1) = p_0^2 + p_1^2 - 2p_0p_1 \hspace{0.1cm}cos(\theta) \ .$$

$\quad$ Expandindo $(p_0 - p_1)^2$ obtemos

$$(p_0 - p_1)^2 = p_0^2 - 2p_0p_1 + p_1^2 \ .$$

$\quad$ Deste modo, podemos reduzir a equação para

$$m_0c(p_0 - p_1) = p_0p_1(1-cos(\theta))$$

ou 

$$\frac{1}{p_1} - \frac{1}{p_0} = \frac{1}{m_0c}(1-cos(\theta)) \ .\tag{2.3}$$

$\quad$ Multiplicando a equação acima por $h$ temos que

$$\Delta \lambda = \lambda_1 - \lambda_0 = \frac{h}{m_0c}(1 - cos(\theta)) \ .\tag{2.4}$$

$\quad$ O termo $\frac{h}{m_0c}$ tem como resultado um valor de $2.43 \times 10^{-12}m = 0.0243A = 2.43pm$. Este valor, por sua vez é chamado de comprimento de Compton. Obtida a equação de Compton é possível observar que o deslocamento de Compton depende apenas do espalhamento $\theta$, e não do comprimento de onda inicial $\lambda_0$.

# 3 - Implementando códigos

Nesse código, observaremos o aumento do comprimento de onda em função do angulo $\theta$.

*Constantes*

```python
c = speed_of_light 
h = Planck
m = electron_mass
```

Utilizando a equação de Compton que deduzimos acima, temos que:

```python
theta = np.arange(0.,1.,1/180)*np.pi # Angulo theta em pi-radianos
delta_lambdaa = 2.43e-12*(1-np.cos(theta)) #equação de Compton
theta_graus = np.arange(0,180) # Angulo theta em graus para plotar no gráfico
print(delta_lambdaa)
```

Deste modo, é notório que o deslocamento Compton varia conforme o ângulo theta varia.

```python
# comprimento de onda do fóton incidente
plt.figure(figsize=(9,6))
plt.style.use('dark_background')
plt.grid('--')
plt.plot(theta_graus,delta_lambdaa,color='red')
plt.title('Gráfico deslocamento Compton em função do angulo de espalhamento')
plt.xlabel('Ângulos (º)')
plt.ylabel([0,6e-12])
plt.ylabel('Deslocamento Compton')
```

$\quad$ O gráfico apresentado acima mostra a relação entre o deslocamento Compton e o ângulo theta. O efeito Compton é uma consequência direta da natureza dual das partículas e das ondas. Este gráfico oferece uma visualização clara da dependência do deslocamento Compton em relação ao ângulo de espalhamento.

$\quad$ Podemos também deixar o gráfico um pouco mais dinâmico fazendo uma animação.

```python
c = speed_of_light 
h = Planck
lambdaa_RaioX = 10e-9
m = electron_mass
theta_graus = np.arange(0,180)

fig, ax = plt.subplots()

x_dados, y_dados = [],[]
ln, = plt.plot([],[],'r-')

def init():
    ax.set_xlim([0,180])
    ax.set_ylim([0,5.5e-12])
    return ln,

def atualizar(frame):
    theta_rad = np.radians(frame)
    lambdaa_linha = 2.43e-12*(1-np.cos(theta_rad)) #equação de Compton
    y_dados.append(lambdaa_linha)
    x_dados.append(frame)
    ln.set_data(x_dados,y_dados)
    return ln,

ani = animation.FuncAnimation(fig,atualizar,frames=theta_graus,init_func=init,blit=True)
ani.save('compton.gif',writer='imagemagick',fps=30)
#Image(filename='compton.gif')
```

## Extra

$\quad$ Antes do fóton se chocar com a partícula, vimos que a energia do fóton é dada por 𝐸=ℎ𝑓. No espalhamento Compton, a energia do fóton após a colisão pode ser calculada pela equação abaixo:

$$E_1 = \frac{E_0}{(\frac{E_0}{mc^2})(1 - cos\theta)+1} \ .$$

$\quad$ Podemos deduzi-lá por conservação de momento e energia, juntamente com princípios da relatividade. Multiplicando a Equação **(2.1)** por $\frac{1}{c}$, temos que:

$$\frac{1}{cp_1} - \frac{1}{cp_0} = \frac{1}{m_0c^2}(1-cos(\theta)) \ .$$ 

$\quad$ Visto que $cp_1$ se trata da energia após a colisão e $cp_0$ se trata da energia antes da colisão, temos que

$$\frac{1}{E_1} - \frac{1}{E_0} = \frac{1}{m_0c^2}(1-cos(\theta)) \ .$$ 

$\quad$ Desenvolvendo a equação

$$\frac{1}{E_1} = \frac{1}{m_0c^2}(1-cos(\theta)) + \frac{1}{E_0}$$ 

$$ \implies 1 = \left[\frac{1}{m_0c^2}(1-cos(\theta)) + \frac{1}{E_0}\right]E_1$$

$$ \implies 1 = \left[\frac{E_1(1-cos(\theta)) + mc^2}{m_0c^2E_0}\right]E_1 \ .$$

$\quad$ Agora, isolando $E_2$ e simplificando a equação chegamos em:

$$E_1 = \frac{E_0}{(\frac{E_1}{mc^2})(1 - cos\theta)+1} \ .$$

$\quad$ No código abaixo, definimos a energia inicial do fóton como $E_1$ e calculamos a energia do fóton após a colisão $E_2$ para cada ângulo de espalhamento. Em seguida, colocamos a relação entre o ângulo de espalhamento e a energia do fóton após a colisão. Isso nos permitirá visualizar como a energia do fóton varia com o ângulo de espalhamento, conforme a equação do Efeito Compton.

```python
f = 10**20 # frequência do raio x em Hz
E1 = h*f
E2 = E1 / (1 + (E1 / (m * c**2)) * (1 - np.cos(theta)))

theta_graus = np.degrees(theta)
```

```python
plt.plot(theta_graus,E2,color='blue')
plt.grid(True)
plt.xlabel('Ângulo de Espalhamento (graus)')
plt.ylabel('Energia do Fóton após a Colisão (J)')
plt.title('Relação entre Ângulo de Espalhamento e Energia do Fóton')
plt.show()
```
