## 1. Sistemas Dinâmicos e Neurociência 

### 1.1 Uma breve introdução à Sistemas Dinâmicos
    
$\quad$ Para começar, precisamos entender alguns conceitos que serão importantes daqui para frente, principalmente no que diz respeito ao estudo dos sistemas dinâmicos e suas aplicações para modelar o funcionamento de neurônios. 

$\quad$ Um sistema dinâmico consiste em um conjunto de variáveis que descrevem seu estado atual e uma ou mais equações diferenciais não-lineares que descrevem a evolução temporal dessas variáveis de estado. Por esse motivo, um sistema dinâmico descreve como um estado em um dado momento depende dos estados em momentos anteriores **[1]**. Para esse estudo, interpretar um conjunto de equações diferenciais não é tão eficiente para compreender um determinado sistema, em razão disto, utiliza-se o **retrato de fase** de um sistema dinâmico para compreender melhor como ele funciona. Um retrato de fase é, em suma, uma descrição do que acontece com os valores das variáveis de estado com o passar do tempo, e com isso conseguimos fazer uma análise qualitativa da dinâmica do sistema sem sequer ser necessário que saibamos todos os detalhes que governam a natureza do mesmo **[1]**. Em um retrato de fase é mostrado certos elementos importantes para determinar a natureza da trajetória de qualquer uma dessas variáveis, como principais para o estudo de sistemas dinâmicos temos: 

- *Os pontos de equilíbrio*: Pontos no retrato de fase para o qual todas as trajetórias são direcionadas (pontos de equilíbrio estável) ou da qual todas as trajetórias se afastam (pontos de equilíbrio instável);
- *Ciclos limite*: Trajetórias na qual as variáveis podem se manter ciclicamente ao longo do tempo; 
- *Bifurcações*: Transições entre dinâmicas diferentes em um mesmo sistema.

Esses são alguns conceitos importantes para a compreensão de retratos de fase de qualquer sistema dinâmico.


### 1.2 Fisiologia básica dos neurônios

$\quad$ Os neurônios são as células fundamentais para o funcionamento neural, sendo responsáveis pela transmissão de sinais elétricos através de sinapses (conexão entre neurônios) e da sua anatomia composta principalmente de dendritos e axônio. O processo de disparo neural começa quando o potencial de membrana é alterado em resposta a estímulos elétricos e químicos, acarretando a geração de um potencial de ação.



Figura 1: Funcionamento de uma bomba de sódio-potássio. *LibreTexts Biology*

$\quad$ Quando tratamos do mecanismo de ativação dos neurônios vemos que, em repouso, o neurônio mantém um potencial de membrana (em torno de $- 70$ $mV$) devido às ações das bombas de sódio-potássio que controlam a concentração de sódio e potássio dentro da célula - alta de sódio do lado de fora, e alta de potássio no de dentro. Ao sofrer um estímulo, o neurônio passa por uma despolarização que torna o potencial de membrana menos negativo. Caso esse processo alcance o valor de um limiar (cerca de $-55 $ $mV$), os canais de sódio-potássio se abrem, fazendo com que seja gerado o potencial de ação e o neurônio dispare. Após esse disparo, o neurônio passa por uma repolarização em que os canais das bombas se abrem possibilitando a saída dos íons de potássio da célula, restaurando o potencial negativo da membrana. Assim, a célula passa por um período em que não pode disparar um potencial de ação imediatamente - período refratário. 

$\quad$ Ou seja, podemos afirmar que os neurônios são excitáveis devido à sua capacidade de gerar potenciais de ação — que são sinais elétricos rápidos e de grande amplitude - em reposta a estímulos. Essa capacidade de gerar e propagar sinais elétricos é uma das características fundamentais dos neurônios e é vital para o funcionamento do sistema nervoso. E é isto que os modelos de ativação neuronal exploram, ajudando a entender os processos de excitação e inibição que regem a atividade neural.  

### 1.3 Neurociência: dinâmica de neurônios

$\quad$ Em 1952, Alan Lloyd Hodgkin e Andrew Huxley descreveram um modelo matemático para a descrição da propagação e inativação dos potenciais de ação de um neurônio. O modelo foi chamado de Modelo Hodgkin-Huxley e é formado por um sistema de quatro equações diferenciais que se conectam por meio da condutância de canais de Sódio, Potássio, etc., que são canais dependentes de voltagem **[2]**. Normalmente, as principais variáveis que descrevem a dinâmica neuronal podem ser classificadas em três classes, de acordo com sua função e escala de tempo: 

- Potencial de membrana.
- Variáveis de excitação, como ativação de uma corrente de Sódio.
- Variáveis de recuperação, como inativação de corrente de Na$^+$ e ativação de corrente rápida K$^+$.

$\quad$ Estas e outras variáveis em um neurônio, ao seguir as equações diferenciais, podem resultar em dinâmicas distintas, por exemplo: tomemos um neurônio que inicialmente dispara apenas uma vez ao ser estimulado, mas ao ser estimulado mais vezes, dispara de forma repetida. Ao estudarmos o retrato de fase desse tipo de dinâmica, poderíamos dizer que, para um estímulo pouco intenso, o sistema tendeu ao equilíbrio (ponto de equilíbrio estável), contudo, para um estímulo mais intenso, o sistema ganha uma trajetória mais alongada e cíclica. O ponto em que temos uma mudança de dinâmica é no caso do neurônio, ao ser estimulado mais vezes, disparar repetidamente, nesse podemos dizer que houve uma bifurcação no sistema e a trajetória de um único disparo se tornou um ciclo limite (disparos seguidos).


## 2. O Oscilador de Van der Pol

$\quad$ O oscilador van der Pol **[3]** **[6]**, denominado como um oscilador de relaxamento, possui a característica de ser estável, porém dissipativo. O oscilador de Van der Pol foi proposto por Balthasar Van der Pol em 1926, com o intuito de modelar circuitos com válvulas, mas desde então vem sendo usado nas mais diversas áreas - incluindo biologia. 
Este oscilador, frequentemente empregado para a investigação das propriedades de osciladores não lineares, se torna muito útil quando desejamos modelar fenômenos oscilatórios, que no nosso caso, será o disparo de neurônios.

A Equação do oscilador de Van der Pol é: 

$$
\frac{d^2x}{dt^2} - \mu(1-x^2)\frac{dx}{dt} + x = 0, \tag{1}
$$

onde $\mu$ mede a força do termo de amortecimento e $x(t)$ é a posição do oscilador.



Figura 2: Retrato de fase do Oscilador de Van der Pol com $\mu = 1$. *Imagem tirada do artigo original de Van der Pol.* **[6]**

### 2.1 Implementando o oscilador de Van der Pol

$\quad$ Para estudarmos o oscilador de Van der Pol neste ``Notebook``, inicialmente precisamos definir a função ``van_der_pol`` que implementa a Equação (1). Assim como definir os parâmetros da equação, no nosso caso o $\mu$ (variável ``mu``) e a posição e velocidade iniciais.


````python
def van_der_pol(t, z, mu):
    x, dxdt = z  
    dzdt = [dxdt, mu * (1 - x**2) * dxdt - x]
    return dzdt

mu = 0 # Quanto maior o valor de mu, mais longe de um círculo fica a figura
z0 = [0.01, 0.01]  # Posição e velocidade iniciais
t = np.linspace(0, 50, 1000)
````

$\quad$ Para fins de comparação com a Figura (2), vamos definir a variável ``mu = 1``. E, usando a função ``solve_ivp``, encontramos a solução da equação tendo as condições iniciais próximas de zero como no artigo. Assim, o gráfico final será o retrato de fase do oscilador de Van der Pol, que pode ser interpretado como um **ciclo limite amortecido**.

````python
mu = 1
sol = solve_ivp(van_der_pol, [0, 50], z0, args=(mu,), t_eval=t)
x = sol.y[0] #posição
y = sol.y[1] #velocidade

plt.figure(figsize=(6,8))
plt.tick_params(direction='in', which = 'both', bottom = 1, top =1, left = 1, right = 1)
plt.title('Oscilador de Van der Pol 2D')
plt.xlabel(f'$x$')
plt.ylabel(f'$dx/dt$')
plt,
plt.plot(x, y, linewidth=0.5, label = f" $\mu$ = {mu}")
plt.legend(frameon=False)
````

$\quad$ Variar o valor do $\mu$ impacta na forma que o retrato de fase toma. Com isso em mente, alteramos a célula anterior com o intuito de melhorar a visualização com os retratos de fase sobrepostos.   

````python
mu = 0
plt.figure(figsize=(6,8))
while mu < 5:
    sol = solve_ivp(van_der_pol, [0, 50], z0, args=(mu,), t_eval=t)
    x = sol.y[0] #posição
    y = sol.y[1] #velocidade
    plt.plot(x, y, lw = 0.5, label = f" $\mu$ = {mu}")
    if mu in [0,4]:
        plt.legend(frameon = False)
    mu += 1


plt.tick_params(direction='in', which = 'both', bottom = 1, top =1, left = 1, right = 1)
plt.scatter(z0[0],z0[1], c = 'black') # Posição inicial
plt.title('Oscilador de Van der Pol 2D')
plt.xlabel(f'$x$')
plt.ylabel(f'$dx/dt$')
plt.show()
````

$\quad$Agora, o código se torna mais simples, pois desejamos criar um gráfico apenas da posição do oscilador ao longo do tempo, e com o gráfico final, observa-se um comportamento distinto de um oscilador normal devido ao termo dissipativo ``mu`` ($\mu$). Se o leitor(a) desejar, pode alterar o valor de ``mu`` e as condições iniciais em ambos os códigos e verificar ligeiras variações no formato do retrato de fase ou do gráfico de posições do oscilador.

````python
mu = 5

z0 = [1.0, 1.0]
sol = solve_ivp(van_der_pol, [0, 50], z0, args=(mu,), t_eval=t)
x = sol.y[0]
y = sol.y[1]

plt.tick_params(direction='in', which = 'both', bottom = 1, top =1, left = 1, right = 1)
plt.title('Van der Pol Oscillator 1D')
plt.xlabel('t')
plt.ylabel('x')
plt.plot(t, x, label='Posição (x)')
````

## 3. O Modelo FitzHugh-Nagumo

$\quad$O modelo de FitzHugh-Nagumo **[4][5]** é uma simplificação do modelo de Hodgkin-Huxley, utilizado para descrever a dinâmica de neurônios e outros sistemas biológicos que exibem comportamentos de excitação, como os impulsos nervosos. Ele foi desenvolvido por Richard FitzHugh (1961) **[4]** e, posteriormente, por J. Nagumo e colaboradores (1962) **[5]**, para capturar as características essenciais da excitação e repolarização neuronal de forma mais simples, sem a alta complexidade do modelo de Hodgkin-Huxley. Este modelo é formado por duas equações diferenciais: 

$$ \frac{dV(t)}{dt}= V - \frac{V^3}{3} - w + I,\tag{2}$$

$$ \frac{dw(t)}{dt}= \epsilon (V + a - bw), \tag{3}$$

onde $V$ é o potencial de membrana do neurônio, $w$ é a variável de recuperação (relativa à inativação dos canais iônicos no neurônio), $I$ é a corrente inicial injetada no sistema (ou estímulo externo de outro neurônio) e os parâmetros a, b e c controlam a dinâmica do sistema (como forma da curva no retrato de fase e mudanças da variável $w$). Vamos tentar implementá-lo e observar o gráfico das duas equações do modelo.

### 3.1 Implementando o modelo FitzHugh-Nagumo em python


$\quad$Inicialmente podemos definir a função ``fitzhugh_nagumo`` que implementa o modelo tratado na seção anterior, como também os parâmetros: $a$, chamado de parâmetro de excitação que afeta a transição entre os disparos e repousos; $b$, o parametro de recuperação que controla como a variável de recuperação ($w$) afeta o potencial de membrana ($V$); e o $\epsilon$ que rege a diferença nas taxas de variação de $w$ e $V$, geralmente com valores $\epsilon \ll 1$ para criar um sistema excitável.

````python
# Parâmetros do modelo de FitzHugh-Nagumo
epsilon = .08 
a = 0.0 
b = 0.0

# Implementar o modelo de FitzHugh-Nagumo

def fitzhugh_nagumo(X, t):
    V, W = X
    dVdt = V - (V**3) / 3 - W + I
    dWdt = epsilon * (V + a - b * W)
    return [dVdt, dWdt]
````

$\quad$ Fazemos o mesmo processo que para o oscilador de Van der Pol, definimos as condições iniciais e corrente externa (estímulo) que estamos considerando constante aqui, e extraímos as soluções.

````python
# Condições Iniciais
V0 = 1.0
W0 = 1.0
X0 = [V0, W0]

# Pontos de tempo para integração
t = np.linspace(0, 150, 1000)  

I = 0.5

sol = odeint(fitzhugh_nagumo, X0, t)

# Extrair as soluções para V e W
V = sol[:, 0]
W = sol[:, 1]
````

$\quad$ Tendo essas variáveis definidas, podemos plotar o gráfico da variação do potencial da membrana com o tempo, mostrando os disparos. E também analisar o comportamento da variável de recuperação com o segundo plot.

````python
plt.plot(t, V, label='V (Voltage)')
plt.title('Modelo FitzHugh-Nagumo')
plt.xlabel('Tempo')
plt.ylabel('Potencial da membrana (V)')
plt.tick_params(direction='in', which = 'both', bottom = 1, top =1, left = 1, right = 1)
plt.show()


plt.plot(t, W, c = 'orange', label=' W (Variável de Recuperação)')
plt.title('Modelo FitzHugh-Nagumo')
plt.xlabel('Tempo')
plt.ylabel('Variável de Recuperação (W)')
plt.show()
````

$\quad$ Nota-se que existe tanto o aumento do potencial de membrana (fase de excitação) quanto a diminuição deste (fase de repolarização). Além disso, a depender dos parâmetros, como no caso, o sistema exibe ciclos de excitação e recuperação, descrevendo um caráter oscilatório esperado para a dinâmica, observada no gráfico das variáveis do modelo ao longo do tempo. Também vale destacar a interação entre o potencial de membrana e a variável de recuperação, em que podemos ver que o valor máximo de $w$ coincide com o mínimo de $V$.

$\quad$ Como é possível observar, o modelo mostra um caráter oscilatório nas variáveis que compõem o modelo, gerando um gráfico muito parecido com o gráfico da posição do oscilador de Van der Pol em uma dimensão. Se criarmos um gráfico da variável de recuperação com o potencial de membrana, teremos, ao final, o retrato de fase do modelo de FitzHugh-Nagumo. Por sua vez, esse retrato de fase mostra a estabilidade que os neurônios atuam comparado com a figura equivalente do oscilador de Van der Pol.


````python
plt.scatter(W[0],V[0], color='black', label='Estado inicial')

plt.plot(V, W, label='Retrato de fase', linewidth = 1)

plt.tick_params(direction='in', which = 'both', bottom = 1, top =1, left = 1, right = 1)
plt.title('Modelo FitzHugh-Nagumo 2D')
plt.xlabel('Potencial de membrana, V')
plt.ylabel('Variável de recuperação, W')
````

## 4. Conclusão

$\quad$ Com o desenvolvimento do trabalho em relação ao estudo e à implementação em Python do oscilador de Van der Pol e do modelo neuronal de FitzHugh-Nagumo, pode-se perceber certas semelhanças entre ambos os modelos de sistemas dinâmicos:

- A equação do oscilador de Van der Pol mostra um termo não linear de amortecimento​ $\mu (1-x^2)\frac{dx}{dt} $, que pode ser comparado ao comportamento da variável de recuperação $w$ no modelo de FitzHugh-Nagumo. 
$\newline$
- Podemos perceber, como já foi mencionado, que ambos os modelos possuem características oscilatórias semelhantes, nos permitindo relacionar o termo não-linear $V$, do modelo de FitzHugh-Nagumo, ao termo não-linear no oscilador de Van der Pol, que envolve uma função do tipo $x^2$.
$\newline$
- O parâmetro $\mu$ no oscilador de Van der Pol controla a intensidade do amortecimento e, portanto, regula a não-linearidade das oscilações. Analogamente, os parâmetros $a$ e $b$ do modelo de ativação neuronal controlam a velocidade da dinâmica de recuperação e a relação entre a excitação e a recuperação.  

$\quad$ Por fim, conclui-se que o modelo de FitzHugh-Nagumo é, de fato, uma adaptação e generalização do Oscilador de Van der Pol, aplicada a disparos neurais. Enquanto o oscilador de Van der Pol foi inicialmente pensado para lidar com circuitos oscilantes, o modelo de Fitzhugh-Nagumo o expande para descrever a dinâmica dos neurônios e elementos fundamentais para a atividade neural.
