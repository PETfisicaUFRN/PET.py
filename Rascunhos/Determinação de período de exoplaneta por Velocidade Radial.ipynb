## 2 Velocidade Radial

### 2.1 Mecânica orbital

$\quad$ Antes de introduzirmos o conceito de velocidade radial, é necessário apresentar uma base conceitual que permita ao leitor compreender plenamente esse elegante método da astronomia observacional. Para isso, consideremos um sistema composto por uma estrela e um planeta. O centro de massa $\vec{R}$ do sistema estrela-planeta é dado por:

$$
\vec{R} = \frac{M_S \cdot \vec{r}_S + M_P \cdot \vec{r}_P}{M_P + M_S}.
\tag{1}
$$

onde $M_S$ e $M_P$ representam as massas da estrela e do planeta, respectivamente, e $\vec{r}_S$ e $\vec{r}_P$ são suas posições relativas.

$\quad$  Ao derivarmos essa equação em relação ao tempo, obtemos a velocidade do centro de massa. Para enxugar a notação, denotemos $M = M_S + M_P$. Assim:

$$
\vec{V} = \frac{M_S \cdot \vec{v}_S + M_P \cdot \vec{v}_P}{M}.
$$

Se multiplicamos essa expressão pela massa total do sistema, obtemos, por sua vez, o momento linear do centro de massa: 

$$
\vec{P} = M\vec{V} = M_S \cdot \vec{v}_S + M_P \cdot \vec{v}_P.
\tag{2}
$$

$\quad$ Como o sistema está isolado, isso significa que não há forças externas atuando sobre a estrela ou o planeta. A única interação presente é a força gravitacional mútua, que, neste contexto, é uma força interna ao sistema. Sendo assim, o momento linear total do sistema (isto é, do centro de massa) é conservado. Isso pode ser representado matematicamente por:

$$
\frac{d\vec{P}}{dt} = \vec{F} = 0 \implies \vec{P} = \vec{k},
$$

onde $\vec{k}$ representa um vetor constante no tempo. Se o momento linear é conservado, é fácil notar que o centro de massa só pode estar sobre movimento retilíneo uniforme. Isso significa que um referencial cuja origem coincide com o centro de massa do sistema constitui um referencial inercial. Equacionando essas informações através de (1) e (2), teremos:

$$
\vec{R} = 0 \implies M_S \cdot \vec{r}_S = - M_P \cdot \vec{r}_P \ , \ \text{e}
\tag{3}
$$
$$
\vec{P} = 0 \implies M_S \cdot \vec{v}_S = - M_P \cdot \vec{v}_P.
\tag{4}
$$

$\quad$ Além disso, a Terceira Lei de Newton afirma que, se um corpo provoca uma força em outro, este reproduz uma reação de mesmo módulo e direção, porém de sentido oposto. Para o sistema estrela-planeta, essa lei pode ser expressa matematicamente como

$$
M_S \cdot \vec{a}_S = - M_P \cdot \vec{a}_P.
\tag{5}
$$

$\quad$ Ademais, importante lembrarmos que a força de interação entre esses corpos é dada pela Lei da Gravitação Universal de Newton:

$$
\vec{F}_{12} = - \frac{G M_1 M_2}{|\vec{r}_1 - \vec{r}_2|^2} \frac{\vec{r}_1 - \vec{r}_2}{|\vec{r}_1 - \vec{r}_2|},
\tag{6}
$$

onde lemos o índice da força como: "a força causada no corpo 1 devido à presença do corpo 2".

$\quad$ O conjunto de expressões (3), (4) e (5) descreve o comportamento dinâmico do sistema, relacionando as grandezas físicas da estrela e do planeta.
Vamos analisar o que essas equações nos revelam até aqui. Ao considerarmos um referencial inercial cuja origem coincide com o centro de massa (também chamado de baricentro), estamos assumindo que esse ponto permanece em repouso ao longo da evolução temporal do sistema. No entanto, essa condição não se aplica individualmente à estrela ou ao planeta. Na verdade, a força gravitacional mútua descrita em (6) faz com que ambos os corpos orbitem o centro de massa, de modo que este permaneça fixo. Isso implica que as posições da estrela e do planeta estão restritas à reta que liga cada corpo ao baricentro — como indicado pela equação (3). A equação (4), por sua vez, mostra que suas velocidades são opostas em sentido, mas possuem a mesma direção. Isso indica que ambos os corpos se movem em órbitas coplanares, com mesmo sentido de rotação ao redor do centro de massa, garantindo a conservação do momento linear total do sistema.

### 2.2 Espectro eletromagnético e Efeito Doppler

$\quad$ Durante o processo de fusão nuclear em seu interior, a estrela gera quantidades impressionantes de energia. Essa energia não se manifesta apenas como calor, mas também na forma de radiação eletromagnética — a ampla faixa do espectro que inclui o que conhecemos como luz visível. A distribuição da radiação emitida por comprimento de onda no núcleo da estrela assemelha-se à curva de radiância espectral de um corpo negro, descrita por Planck por meio da seguinte equação:

$$
R_T(\lambda) = \frac{2hc^2}{\lambda^5} \cdot \frac{1}{e^{\frac{hc}{\lambda k T}} - 1}.
$$

$\quad$ Nessa expressão, $R_T(\lambda) d\lambda$ representa a energia irradiada por unidade de tempo, entre os comprimentos de onda $\lambda$ e $\lambda + d\lambda$, por unidade de área de uma superfícia a temperatura absoluta $T$. Para os leitores que desejarem se aprofundar nesse tema, recomendamos a referência [6]. A radiação de corpo negro descreve como um objeto idealizado (o corpo negro) emite radiação em um espectro contínuo, ou seja, cobrindo todos os comprimentos de onda. A forma da curva de radiância espectral depende da temperatura do corpo emissor: quanto maior a temperatura, mais intensa e deslocada para comprimentos de onda menores será a radiação. A imagem a seguir ilustra a curva de radiância para diferentes temperaturas: 



Gráficos para radiância espectral de corpo negro para diferentes temperaturas. Fonte: [https://pt.wikipedia.org/wiki/Lei_de_Planck](https://pt.wikipedia.org/wiki/Lei_de_Planck)

$\quad$ Entretanto, durante o percurso da radiação eletromagnética do núcleo até a superfície da estrela, determinados comprimentos de onda são absorvidos — ou seja, “desaparecem” do espectro observado. Esse fenômeno ocorre porque, nas camadas mais externas da estrela, há átomos em estados discretos de energia. Quando os fótons emitidos no núcleo atravessam essas regiões, parte deles tem sua energia absorvida por esses átomos, provocando uma transição eletrônica para um nível de energia mais alto. Como consequência, esses comprimentos de onda não chegam até nós, e observamos, no espectro da estrela, linhas escuras características — chamadas de linhas de absorção. Portanto, a radiação observada não forma um espectro completamente contínuo. Cada linha de absorção é específica de um elemento químico, funcionando como uma “impressão digital” que permite identificar a composição da atmosfera estelar. Com base nessas linhas, podemos inferir os elementos presentes e, assim, reconstruir o espectro esperado da estrela. A imagem a seguir mostra o espectro eletromagnético do Sol na faixa do visível, evidenciando suas linhas de absorção.



Espectro eletromagnético do Sol. Fonte: [https://spacetoday.com.br/o-espectro-do-sol/](https://spacetoday.com.br/o-espectro-do-sol/)

$\quad$ Agora, tratemos de um fenômeno fundamental no estudo das ondas: o Efeito Doppler. Para introduzi-lo qualitativamente, consideremos o exemplo de uma onda sonora. Imagine que você esteja parado na calçada em um dia tranquilo e, ao longe, ouve a sirene de um carro de polícia se aproximando. À medida que o veículo se aproxima, o som parece mais agudo; ao se afastar, torna-se mais grave. Embora a intensidade sonora também varie com a distância, ela não é o foco aqui. O que nos interessa é a mudança na frequência percebida. Se tanto você quanto a fonte sonora estiverem em repouso, a frequência será a mesma, independentemente da posição. Logo, essa variação não pode depender apenas da posição relativa, mas sim da velocidade relativa entre o observador e a fonte. Quando a fonte se aproxima, a frequência percebida aumenta; quando se afasta, diminui.

$\quad$ Esse fenômeno não é exclusivo de ondas mecânicas — ele também ocorre com ondas eletromagnéticas, como a luz. Foi justamente observando o desvio espectral da luz de galáxias distantes que Edwin Hubble concluiu que o universo está em expansão. Quando uma fonte de radiação eletromagnética se afasta de um observador, sua frequência diminui (e o comprimento de onda aumenta), o que chamamos de *redshift* (desvio para o vermelho).Inversamente, quando a fonte se aproxima, ocorre o *blueshift* (desvio para o azul).

$\quad$ Para generalizar esse fenômeno, consideremos agora um cenário relativístico (Para uma introdução à Relatividade Restrita, sugerimos a referência [7]). Assim, imaginamos que a fonte possa se mover até mesmo em velocidades próximas a da luz. Suponha uma construção geométrica no plano cartesiano bidimensional, com eixos $x$ e $y$. Colocamos um observador na origem, e uma fonte de luz (por exemplo, verde) posicionada sobre o eixo $x$. A fonte move-se com uma velocidade $v$, formando um ângulo $\theta$ com o eixo $x$ - ou seja, o ângulo entre a direção da velocidade da fonte e a linha que conecta a fonte ao observador.. O comprimento de onda da luz emitida pela fonte, ao ser observada, terá seu valor dado por:

$$
\nu_{obs} = \nu_0 \frac{(1+ \beta \cos{\theta})}{\sqrt{1 - \beta^2}} \ , \ \beta = \frac{v}{c}.
\tag{7}
$$

Nessa expressão, $\nu_{obs}$ se refere à frequência da onda detectada pelo observador e $\nu_0$ à frequência originalmente emitido pela fonte. Porém, se a velocidade da fonte for muito inferior a velocidade da luz, ou seja $v \ll c$, conseguimos escrever (7) de uma maneira bem mais agradável:

$$
\nu_{obs} = \nu_0 (1+ \beta \cos{\theta}).
\tag{8}
$$

$\quad$ Note que, se a fonte de luz verde se aproxima do observador, logo $\theta = 0° \implies \cos{\theta} = 1$ e, portanto, a frequência da onda aumenta, fazendo com que a luz se torne mais azulada. Se $\theta = 180° \implies \cos{\theta} = -1$ e a frequência diminui, deixando-a mais vermelha. Agora, se $\theta = 90° \implies \cos{\theta} = 0$, logo $\nu_{obs} = \nu_{0} \ $, não existindo o deslocamento Doppler.

### 2.3 Determinação da velocidade radial

$\quad$ Agora que já desenvolvemos os conceitos fundamentais, podemos finalmente compreender como os astrônomos determinam a velocidade radial de uma estrela a partir de seu espectro eletromagnético. Como discutido na seção 2.1, em um sistema formado por uma estrela e um planeta, ambos orbitam o centro de massa (baricentro) do sistema, com aceleração dirigida para esse ponto e em um mesmo sentido de rotação (horário ou anti-horário). Além disso, mostramos que a posição relativa entre os corpos é tal que o planeta, a estrela e o baricentro estão alinhados ao longo de uma reta, com vetores de posição opostos em relação ao centro de massa.

$\quad$ No entanto, ao orbitar o baricentro, a estrela apresenta velocidade não nula. Se parte desse vetor velocidade estiver orientado na direção da linha de visada do telescópio, ocorrerá o efeito Doppler, o que levará a uma modificação observável no espectro da estrela. Esse resultado é notável, pois contradiz a ideia intuitiva de que as estrelas seriam corpos estacionários em seus sistemas. Na verdade, conseguimos detectar com alta precisão esse pequeno "bamboleio" da estrela, causado pela presença de um planeta em órbita. Se não houvesse nenhum planeta interagindo com a estrela, o centro de massa do sistema coincidiria aproximadamente com o centro geométrico da própria estrela. (Ressaltamos, no entanto, que outros fatores também podem causar o deslocamento do baricentro, como a presença de uma segunda estrela — um sistema binário. Esse caso, porém, não será abordado aqui, pois foge ao escopo deste estudo.)

$\quad$ Para medir a velocidade radial, os astrônomos primeiro realizam um estudo detalhado da estrela, determinando parâmetros como sua classe espectral, temperatura, e outras propriedades físicas relevantes. Com essas informações, é possível prever a forma esperada do seu espectro eletromagnético. Utilizando um espectrômetro — instrumento capaz de decompor a luz da estrela em seus comprimentos de onda constituintes — os cientistas analisam as linhas escuras (linhas de absorção) do espectro, previamente discutidas. Ao longo do tempo, essas linhas são monitoradas. Caso a estrela esteja periodicamente se aproximando e se afastando do observador, essas linhas sofrerão sucessivos desvios para o azul e para o vermelho, respectivamente. Como as posições dessas linhas de absorção são bem conhecidas, qualquer desvio sistemático pode ser atribuído com segurança ao efeito Doppler. Manipulando a equação (8), podemos obter uma expressão para a velocidade radial da estrela:

$$
v_r = c \frac{\Delta \lambda}{\lambda_0} \ , \ \Delta \lambda = \lambda_{obs} - \lambda_0.
\tag{9}
$$

Nessa expressão, $\lambda_0$ é o comprimento de onda original da linha de absorção (em repouso), e $\lambda_{obs}$ é o comprimento de onda observado. A variável $\lambda$ está associada à frequência $\nu$ via a relação $c = \lambda \ \nu$. Portanto, podemos determinar a velocidade radial da estrela a partir da variação do comprimento de onda de uma determinada linha escura bem conhecida da estrela. Utilizamos a aproximação em que $v \ll c$ pois, geralmente, o movimento das estrelas em torno do baricentro do sistema não é próximo da velocidade da luz.

$\quad$ Uma vez obtidos os valores de $v_r$ em diferentes instantes, é possível construir um gráfico da forma: $v_r \times t$, ou seja, velocidade radial em função do tempo. O objetivo dessa análise é identificar padrões periódicos na variação da velocidade, pois essa periodicidade é uma assinatura direta da presença de um exoplaneta em órbita da estrela. O período observado nesse gráfico corresponde exatamente ao período orbital do planeta, permitindo estimativas confiáveis dos parâmetros orbitais. Buscaremos justificar esse ponto a seguir.

$\quad$ Para simplificar a análise, consideremos que o plano orbital do planeta seja paralelo à linha de visada do observador — ou seja, temos uma “visão total” do movimento relativo entre estrela e planeta. (Na prática, basta que exista algum componente da órbita na direção da linha de visada para que o método de velocidade radial funcione.) Assumimos ainda que ambas as órbitas sejam elípticas. O leitor deve recordar que, em qualquer movimento curvilíneo, a velocidade do corpo é sempre tangente à trajetória — neste caso, uma elipse. A figura a seguir ilustra a geometria da órbita no contexto dessa análise:



Representação da órbita de um sistema estrela-planeta, em torno do seu centro de massa. $t_1$, $t_2$ e $t_3$ representam tempos diferentes, o símbolo $\star$ se refere à estrela e o índice $p$ ao planeta. Fonte: Perryman, M. *The Exoplanet Handbook*. 2 ed. Cambridge. 2018.

$\quad$ Consideremos que a linha de visada do observador coincida com a linha tracejada mostrada na figura anterior. Retomando os conceitos apresentados na seção 2.1: se o planeta está se aproximando do centro de massa, então, pela equação (4), a estrela também deve estar se aproximando. Isso ocorre porque, embora os vetores velocidade tenham sentidos opostos, os corpos estão localizados em lados opostos do baricentro, o que confere uma simetria dinâmica ao sistema. Portanto, sempre que o planeta possui uma componente de velocidade ao longo da linha de visada, a estrela também apresentará tal componente — mas em sentido oposto. É essa componente da velocidade estelar que nos permite realizar a medição da velocidade radial.

$\quad$ Vamos reforçar, agora, a interpretação da equação (3). Essa equação nos mostra que o planeta e a estrela sempre permanecem em lados opostos do centro de massa, o que implica que suas posições estão vinculadas por uma reta imaginária que conecta os dois corpos ao baricentro. Essa configuração impede que o planeta complete dois períodos orbitais enquanto a estrela realiza apenas um — ou seja, seus movimentos são sincronizados. Assim, o período orbital da estrela é exatamente o mesmo do planeta.

$\quad$ Mas por que o período da velocidade radial da estrela também corresponde ao período de sua órbita? A resposta está na forma da curva observada:
A velocidade radial da estrela é nula duas vezes por ciclo orbital — quando ela atinge o pericentro (ponto mais próximo do baricentro) e o apocentro (ponto mais distante). Se realizarmos uma medição contínua da velocidade radial a partir do pericentro, observaremos a seguinte sequência: A velocidade inicia em zero (pericentro) $\rightarrow$ Torna-se positiva (estrela se aproximando) $\rightarrow$ Zera novamente no apocentro $\rightarrow$ Torna-se negativa (estrela se afastando) $\rightarrow$ E retorna a zero no pericentro. 

$\quad$Esse comportamento resulta em uma curva senoidal — típica em gráficos de $v_r \times t$. O intervalo de tempo entre dois pontos idênticos na curva (por exemplo, dois pericentros consecutivos) define o período da oscilação. Para as nossas medidas, o período se refere justamente ao tempo em que as medidas se anularam no pericentro. Entretanto, esse é justamente o tempo necessário para que a estrela dê uma volta completa em torno da sua estrela!

$\quad$ É com base nesse princípio que os astrônomos identificam exoplanetas: ao observar uma curva periódica de velocidade radial, conclui-se que a estrela está orbitando o baricentro do sistema sob a influência gravitacional de um planeta. Assim, o planeta também orbita o centro de massa, com o mesmo período orbital da estrela. Em resumo, é graças ao "bamboleio" da estrela, induzido pela presença de um planeta, que conseguimos detectar a existência desse corpo celeste. ### 3.1 Análise de Fourier

$\quad$ Antes de nos aprofundarmos no método de Lomb-Scargle, é útil introduzir brevemente a Transformada de Fourier e o conceito de periodogramas. Suponha uma função qualquer $g(t)$, absolutamente integrável (isto é, cuja integral do valor absoluto sobre todo o domínio é finita) e com um número finito de descontinuidades do primeiro tipo — também chamadas de descontinuidades de salto. A Transformada de Fourier dessa função é definida por:

$$
\mathcal{F}\{g (t)\}(f) = \int_{-\infty}^{\infty}g(t) e^{-2 \pi i f t} \ dt.
$$

A expressão $\mathcal{F}\{g (t)\}(f)$ representa a transformada de Fourier da função $g(t)$, originalmente no domínio temporal, e resulta em uma função no domínio das frequências, $f$. Aqui, $i$ é a unidade imaginária. A principal ideia por trás da transformada de Fourier é decompor um sinal arbitrário em uma combinação (potencialmente infinita) de funções periódicas de diferentes frequências. A partir dessa transformada, definimos a densidade espectral de potência (PSD, Power Spectrum Density) como:

$$
\mathcal{P}_g(f) = |\mathcal{F}\{g\}|^2.
$$

A função $\mathcal{P}_g(f)$ é real, positiva, e quantifica diretamente a contribuição de cada frequência à composição do sinal original. Em outras palavras, o PSD nos diz “quanta potência” do sinal está associada a cada frequência componente.

$\quad$ Até aqui, tratamos do caso contínuo, no qual assumimos que os dados do sinal estão disponíveis de forma ininterrupta ao longo do tempo. No entanto, essa suposição raramente se sustenta na prática — especialmente em áreas como a astronomia, onde as observações são feitas em momentos discretos e espaçados. Nesse contexto, é necessário recorrer à Transformada Discreta de Fourier (DFT), cuja formulação é dada por:

$$
\hat{g}_k = \sum_{n=1}^N g_n e^{-2\pi i k n /N}.
\tag{10}
$$

Nessa expressão, $\hat{g}_k$ representa o coeficiente da transformada para a frequência discreta $k$, $g_n$ é o valor da amostra do sinal no instante $n$, e $N$ é o número total de amostras. A frequência discreta $k$ está relacionada com a frequência física $f$ por meio da seguinte relação:

$$
f = \frac{k}{N}f_s \ ,
$$

onde $f_s$ é a frequência de amostragem, isto é, a taxa com que os dados foram coletados (em amostras por segundo). É importante destacar que a DFT é capaz de representar apenas frequências no intervalo de $0$ até $f_s/2$, devido ao chamado limite de Nyquist. Frequências acima desse valor se tornam indistinguíveis das mais baixas, causando o fenômeno conhecido como *aliasing*.[9]

$\quad$ Agora que conhecemos a Transformada de Fourier e sua versão discreta, é apropriado introduzir uma ferramenta fundamental para a análise espectral de sinais: o periodograma clássico, também conhecido como Periodograma de Schuster. Sua formulação é dada por:

$$
P_S(f) = \frac{1}{N}\left|\sum_{n=0}^N g_n e^{-2 \pi i f t_n} \right|^2 =  \frac{1}{N}\left[\left(\sum_{n=0}^N g_n \cos{(2 \pi f t_n)}\right)^2 + \left(\sum_{n=0}^N g_n \sin{(2 \pi f t_n)}\right)^2 \right]\ ,
\tag{11}
$$

Aqui, $t_n$ indica o instante de tempo em que a n-ésima amostra do sinal $g$ foi registrada. O periodograma pode ser interpretado como um estimador discreto da densidade espectral de potência (PSD) discutida anteriormente. Para entender melhor sua aplicação, considere a observação de uma estrela. Embora o astro emita luz de forma contínua, as medições que realizamos a partir da Terra são feitas de maneira esparsa e discreta, limitadas por fatores como o ciclo dia-noite, condições atmosféricas ou disponibilidade de instrumentos. Portanto, o sinal contínuo é amostrado de forma incompleta.
Se desejarmos analisar as componentes harmônicas presentes nesse sinal, recorremos ao periodograma. Ele nos fornece uma estimativa de quais frequências estão presentes, mesmo que não reproduza perfeitamente o PSD da fonte contínua. Essa limitação, no entanto, é imposta pelas restrições observacionais e instrumentais — e dentro dessas condições, o periodograma de Schuster oferece a melhor aproximação viável.

### 3.2 Lomb-Scargle

$\quad$ Nos casos em que as amostras de um sinal são obtidas de forma uniformemente espaçada no tempo, o periodograma clássico apresentado na equação (11) é capaz de identificar com precisão as principais frequências presentes no sinal. No entanto, esse cenário ideal raramente se verifica em observações astronômicas. Fatores como condições climáticas, restrições operacionais e a própria rotação da Terra frequentemente impedem a aquisição regular de dados. Além disso, muitas análises exigem a combinação de observações realizadas em diferentes épocas, o que naturalmente introduz espaçamentos temporais irregulares.

$\quad$ Diante dessa limitação, surgiu a necessidade de um método capaz de identificar periodicidades mesmo quando os dados não são uniformemente amostrados. Essa foi a motivação por trás do trabalho de Lomb (1976) [5], que propôs uma abordagem alternativa baseada no método dos mínimos quadrados, um procedimento amplamente utilizado em estatística. Lomb partiu da suposição de que o sinal pode ser aproximado por uma combinação linear de funções trigonométricas. Assim, modelou cada observação $g_n$ do sinal como:

$$
g_n = a \cos{(2 \pi f t_n)} + b \sin{(2 \pi f t_n)},
$$

onde os coeficientes $a$ e $b$ são parâmetros desconhecidos a serem ajustados, e $f$ representa a frequência a ser testada.

$\quad$ Ao aplicar o método dos mínimos quadrados para ajustar esse modelo aos dados, Lomb obteve, após um desenvolvimento matemático rigoroso, uma versão modificada da equação que melhora a estabilidade do ajuste. Essa modificação consiste em introduzir um deslocamento temporal $\tau$ no argumento das funções trigonométricas, de modo que o modelo passa a ser:

$$
g_n = a \cos{(2 \pi f (t_n - \tau))} + b \sin{(2 \pi f (t_n - \tau))}.
$$

O parâmetro $\tau$ é  escolhido de forma que a seguinte condição seja satisfeita:

$$
\sum_{i=1}^N \cos{(2 \pi f (t_n - \tau))}\sin{(2 \pi f (t_n - \tau))} = 0,
$$

ou seja, as componentes cosseno e seno tornam-se ortogonais.

$\quad$ Com essa escolha, a minimização da soma dos quadrados resulta em uma expressão que lembra fortemente a estrutura do Periodograma de Schuster:

$$
\Delta R(f) = \frac{\left[ \sum_n g_n \cos( 2 \pi f (t_n - \tau) \right]^2)}
     {\sum_n \cos^2 (2 \pi f (t_n - \tau))} 
+ 
\frac{\left[ \sum_n g_n \sin (2 \pi f (t_n - \tau)) \right]^2}
     {\sum_n \sin^2( 2 \pi f (t_n - \tau))} .
$$

$\quad$ Esta é, essencialmente, a fórmula que define o Periodograma de Lomb-Scargle em sua forma básica. Antes de apresentarmos sua versão completa e normalizada — que inclui correções estatísticas e melhorias introduzidas por Scargle —, é útil revisar o método utilizado por ele, que levou de maneira independente à mesma formulação.

$\quad$ Em 1982, Scargle [4] abordou o mesmo problema discutido anteriormente: o periodograma clássico funciona adequadamente apenas quando os dados estão uniformemente espaçados. Mas o que fazer quando essa condição não é satisfeita, como ocorre com frequência em observações astronômicas? Diferentemente de Lomb, que utilizou o método dos mínimos quadrados, Scargle adotou uma abordagem estatística, propondo uma modificação do periodograma clássico que satisfizesse certos critérios desejáveis. A definição introduzida por Scargle, amplamente utilizada atualmente, também aparece na referência [9]. Partindo da generalização do periodograma clássico, ele propôs a seguinte forma intermediária:

$$
P(f) = \frac{A^2}{2}\left( \sum_n g_n \cos( 2 \pi f (t_n - \tau) \right)^2 + \frac{B^2}{2}\left( \sum_n g_n \sin( 2 \pi f (t_n - \tau) \right)^2.
$$

Em seguida, demonstrou que, ao escolher os parâmetros $A$, $B$ e $\tau$ de maneira apropriada, seria possível garantir que o periodograma modificado satisfizesse três propriedades fundamentais:

- Reduz-se à forma clássica quando os dados são uniformemente espaçados;
- Possui uma distribuição estatística analiticamente conhecida;
- É invariante sob translações globais no tempo (isto é, deslocamentos do conjunto de tempos observados por uma constante).

A partir dessa construção, e com base também na formulação de Lomb, obteve-se a forma final e generalizada do Periodograma de Lomb-Scargle, dada por:

$$
P_{LS}(f) = \frac{1}{2} \left\{ 
\frac{\left[ \sum_n g_n \cos (2 \pi f (t_n - \tau)) \right]^2}
     {\sum_n \cos^2 (2 \pi f (t_n - \tau))} 
+ 
\frac{\left[ \sum_n g_n \sin (2 \pi f (t_n - \tau)) \right]^2}
     {\sum_n \sin^2 (2 \pi f (t_n - \tau))} 
\right\}.
$$

Nesta expressão, o parâmetro $\tau$ é definido como:

$$
\tau = \frac{1}{4 \pi f} \tan^{-1}\left(\frac{\sum_n \sin (4\pi f t_n)}{\sum_n \cos (4\pi f t_n)} \right).
$$

$\quad$ Com isso, chegamos à formulação de um periodograma robusto e amplamente aplicável a séries temporais com espaçamento irregular: o Periodograma de Lomb-Scargle. Ao contrário do periodograma clássico, essa abordagem é especialmente eficaz em contextos onde os dados são adquiridos de forma não uniforme — como é frequentemente o caso em observações astronômicas. Um ponto notável é que todos os parâmetros que diferenciam o Lomb-Scargle do periodograma tradicional — como o deslocamento temporal $\tau$ — são expressos diretamente em função das frequências analisadas e dos tempos de observação. Isso torna o método não apenas matematicamente consistente, mas também computacionalmente viável e facilmente interpretável.

$\quad$ Graças a essas vantagens, o periodograma de Lomb-Scargle ganhou grande notoriedade na literatura científica. Hoje, é considerado uma das ferramentas mais eficazes para a identificação de periodicidades em dados observacionais, sendo amplamente utilizado não apenas em astronomia, mas também em áreas como climatologia, geofísica, medicina e economia. A seguir, exploraremos a aplicação prática desse método em um sistema planetário real, demonstrando sua utilidade na detecção de exoplanetas a partir da variação na velocidade radial de suas estrelas hospedeiras.

## 4. Determinação do período

$\quad$ Para investigar a existência de periodicidade na curva de velocidade radial da estrela, é necessário, primeiramente, obter os dados processados da velocidade radial em função do tempo. Neste notebook, os dados foram extraídos da plataforma DACE (Data & Analysis Center for Exoplanets) [10], da Universidade de Genebra. Para realizar o download, siga os passos abaixo:

- Ao centro-esquerdo, identifique a aba com o nome "*Radial Velocity Surveys*" e selecione-a

A partir desse conjunto de dados — uma série temporal da velocidade radial — realizaremos a análise para determinar o período orbital de um possível exoplaneta.

$\quad$ Para dar continuidade, primeiro identificamos o diretório onde os arquivos extraídos da plataforma DACE foram salvos. Utilizamos a função `glob.glob()` da biblioteca `glob` para localizar todos os arquivos com extensão `.rdb`, que correspondem às séries temporais exportadas. Esse tipo de arquivo contém os dados de velocidade radial organizados em colunas separadas por espaços ou tabulações.

```python
# Identificamos o diretório onde os arquivos extraídos foram salvos.
pasta = r"C:\Users\José Emanuel\Notebook\51peg"

# Utilizamos a função glob.glob() para localizar todos os arquivos com extensão .rdb, 
# que é o formato padrão fornecido pela plataforma DACE.
arquivos = glob.glob(pasta + r"\*.rdb")
```

$\quad$ Inicializamos 2 vetores nulos que simbolizam o pacote de dados. Utilizamos 2, pois os dados com nome *"ELODIE"* possuem os dados transladados, ou seja, não estão no mesmo nível dos demais (Se o leitor quiser conferir, ao selecionar a estrela *51 Peg*, vá na aba *"Summary"*. Você irá notar uma média muito destoante).

```python
dfs = []
dfs_elodie = []
```

$\quad$ Utilizamos um laço para procurarmos em nossos arquivos, justamente aqueles que possuem uma média de dados muito afastada.

```python
for arq in arquivos:
    # Utilizamos a função read_csv da biblioteca pandas para ler os arquivos no formato .rdb.
    # Apesar da extensão, esses arquivos consistem em tabelas delimitadas por tabulações.
    df = pd.read_csv(arq, sep="\t", skiprows=[1])

    # Acessamos a coluna 'vrad', que representa a velocidade radial. 
    # Com base na média dessa coluna, classificamos os dados conforme sua origem.
    # Arquivos com média superior a -20000 pertencem a um conjunto; os demais, a outro (Elodie).
    if df["vrad"].mean() > -20000:
        dfs.append(df)
    else:
        dfs_elodie.append(df)
```

$\quad$ Após importar os conjuntos de dados desejados, unimos todos em um único DataFrame (estrutura de dados), que concentra as observações utilizadas na análise. Em seguida, calculamos a média da coluna vrad, que armazena os valores da velocidade radial da estrela — essa média poderá ser útil, por exemplo, para normalizar os dados.

```python
df_total = pd.concat(dfs, ignore_index=True)
media = df_total["vrad"].mean()
```

$\quad$ Agora, iremos converter os dados *"ELODIE"*, trazendo a média de seus dados para o mesmo nível dos dados do dataframe *'df_total'*. Começamos iniciando um vetor nulo.

```python
dfs_elodie_convertidos = []
```

$\quad$ Fazemos um laço, percorrendo todos os dados *"ELODIE"*.

```python
for df in dfs_elodie:
    # Tiramos a sua média
    media_elodie = df["vrad"].mean()
    # Trazemos cada dado para a média de dados normalizado
    df["vrad"] = df["vrad"] - media_elodie + media
    # Juntamos todos esses dados no vetor apropriado
    dfs_elodie_convertidos.append(df)
```

$\quad$ Finalmente, juntamos todas as informações em um único dataframe, devidamente normalizado e estruturado.

```python
df_total = pd.concat([df_total] + dfs_elodie_convertidos, ignore_index=True)
```

$\quad$ A seguir, exibimos as primeiras linhas do nosso DataFrame, que mostram as colunas disponíveis no conjunto de dados. Algumas dessas grandezas, no entanto, estão presentes apenas em observações específicas. Nas demais, o valor correspondente aparece como *'NaN' (Not a Number)*, indicando a ausência daquela medição para o respectivo registro.

```python
print(df_total.head())
# Também mostramos a quantidade total de medições obtidas 
print(f"\nTotal de linhas: {len(df_total)}")
```

$\quad$ Neste trabalho, as únicas grandezas que utilizaremos são: a velocidade radial, o erro associado a essa medição e o instante em que a observação foi realizada, registrado em dias julianos — unidade padrão de tempo em astronomia. Para facilitar o tratamento posterior, extraímos essas três informações em variáveis separadas (arrays), acessando diretamente as respectivas colunas do DataFrame.

```python
vrad = df_total["vrad"]
rjd = df_total["rjd"]
erro_vrad = df_total["svrad"]
```

$\quad$ Agora, iremos tocar num ponto muito sutil das técnicas observacionais. Os dados extraídos dos arquivos estão em unidades de tempo conhecida como *Reduced Julian Date* (Data Juliana Reduzida), ou RJD; uma medida astronômica muito utilizada para padronizar o cálculo de datas e eventos temporais. Ela é derivada da *Julian Date* (Data Juliana), ou JD; que é uma medida de tempo contínua de dias. Define-se a Data Juliana como o número do dia Juliano do meio dia anterior, somado com a fração de dia desde então, até a medição. O dia Juliano 0 começou ao meio-dia do dia 1° de Janeiro de 4713 a.C. O dia 1, por sua vez, começou a partir das 12h do dia 2 de Janeiro de 4713 a.C, e assim sucessivamente. Como a Data Juliana possui números muito grandes, convém modificá-las. É justamente isso que RJD faz. Define-se a Data Juliana Reduzida como [2]: 

$$
\text{RJD} = \text{JD} - 2400000.0
$$

$\quad$ Entretanto, essas duas últimas grandezas não levam em conta a velocidade finita da luz e efeitos relativísticos, por exemplo. Assim, para se obter melhor precisão nos dados, se faz necessário transformar essas escalas de tempo. Nesse `Notebook`, utilizaremos a escala conhecida como *Barycentric Julian Date* (Data Juliana Baricêntrica), ou BJD. Ela é uma medida que corrige as possíveis distorções temporais. Aqui, utilizaremos funções prontas da biblioteca `astropy` para fazer essa transformação, mas, se o leitor tiver interesse, a referência [2] disserta muito bem sobre esse tema. Agora, vamos ao que interessa. Primeiro, transformamos os dados de JD para RJD. 

```python
jd = rjd + 2400000.0
```

$\quad$ Precisamos agora das coordenadas astronômicas da estrela e da localização do instrumento utilizado nas medições. Como as observações foram realizadas por diferentes espectrógrafos, adotaremos a localização daquele com o maior conjunto de dados: o *Hamilton Echelle Spectrograph*, localizado no *Lick Observatory*, na Califórnia, EUA.

```python
coord = SkyCoord("22h57m27.98s", "+20d46m07.8s", frame='icrs')
location = EarthLocation.of_site('lick')
```

$\quad$ Utilizamos algumas funções da `astropy` para fazermos a conversão por completo.

```python
# Cria objeto Time com os JD
t_jd = Time(jd, format='jd', scale='utc', location=location)

# Converte para BJD (baricentric) com correção TDB
t_bjd = t_jd.light_travel_time(coord)

# BJD_TDB é JD + tempo de viagem da luz
bjd0 = t_jd.tdb + t_bjd
```

$\quad$ Em seguida, organizamos os dados em um vetor que representa o BJD. Além disso, para tornar os valores semelhantes aos apresentados na plataforma DACE, subtraímos 2.450.000 dos valores de BJD, obtendo assim uma espécie de BJD reduzido. Abaixo, mostramos alguns exemplos desses dados processados.

```python
bjd =(bjd0.value)  # Array com os BJD ;
bjd = bjd - 2450000
bjd
```

$\quad$ Finalmente, fazemos um gráfico de $R_v \times \text{BJD - 2450000.0}$, ou seja, velocidade radial em função do tempo. Com isso, podemos exergar como os nossos dados são distribuídos.

```python
plt.figure(figsize=(10, 5))
plt.plot(bjd,vrad,'o')
plt.ylabel("$R_v$ (m/s)")
plt.xlabel("BJD - 2,450,000 (dias)")
plt.title("Gráfico da velocidade radial da estrela em função do tempo")
plt.grid(True)
plt.show()
```

$\quad$ Com a série temporal devidamente construída, utilizamos o periodograma de Lomb-Scargle, conforme descrito na Seção 3 deste notebook. A biblioteca `astropy` disponibiliza uma função que calcula o periodograma automaticamente com base em três parâmetros: tempo, velocidade radial e erro associado. Além disso, é possível restringir o domínio de frequências em que a análise será realizada. Como já temos um conhecimento prévio sobre o período esperado, limitamos a faixa de frequências consideradas.

```python
ls = astro.LombScargle(bjd,vrad,erro_vrad)
frequencies, power = ls.autopower(
    minimum_frequency = 1/20,
    maximum_frequency = 1/0.5,
    samples_per_peak = 20
)
```

$\quad$ A seguir, apresentamos o gráfico resultante do cálculo do periodograma. Para isso, utilizamos o inverso dos valores de frequência, de modo a representar a potência do método de Lomb-Scargle em função do período, e não da frequência. Essa modificação é adotada porque, geralmente, descreve-se a órbita de um planeta com base em seu período orbital.

```python
plt.figure(figsize=(10, 5))
plt.plot(1/frequencies, power)
plt.xlabel("Período (dias)")
plt.ylabel("Potência Lomb-Scargle")
plt.title("Periodograma de Lomb-Scargle")
plt.xlim(0.5,20)
plt.grid(True)
plt.show()
```

$\quad$ Observamos facilmente que o resultado do periodograma apresenta um pico de alta potência próximo a 5 dias. Devido à intensidade desse pico, o sinal pode ser considerado inequivocamente periódico. Para determinar o valor exato do período, identificamos no vetor de frequências aquela que corresponde ao índice do maior valor no vetor *power*. Em seguida, invertemos essa frequência, obtendo assim o período dominante da amostra.

```python
frequencia_pico = frequencies[np.argmax(power)]
periodo_dominante= 1 / frequencia_pico
periodo_dominante
```

$\quad$ Também estamos interessados em estimar o erro associado à medida do período. Para isso, utilizamos a equação descrita na referência [11]:

$$
\delta P = \frac{\delta \nu  P^2}{2},
$$

onde $\delta P$ representa a incerteza do período dominante $P$, e $\delta \nu$ é a incerteza da frequência, relacionada ao tempo total de observação pela expressão $\delta \nu = 1/T$, com $T$ sendo o intervalo total da campanha observacional. Essa fórmula é válida para conjuntos de dados que não sejam fortemente não uniformes — o que se aplica ao nosso caso. Para outra aplicação dessa equação em estudos de exoplanetas, sugerimos a referência [12].

$\quad$ Calculamos o erro com a expressão descrita acima:

```python
T = bjd.max() - bjd.min()
df = 1/T
dp = df*(periodo_dominante**2)/2
```

$\quad$ Por fim, mostramos o resultado para o período dominante em nosso sinal e o erro associado à essa medida.

```python
print(f"Período = {periodo_dominante:.5f} ± {dp:.5f} dias")
```

$\quad$ Podemos agora utilizar as informações obtidas para avaliar a significância do resultado encontrado no periodograma. Para isso, traçamos uma linha indicando o valor do pico identificado. Além disso, consideramos como intervalo de confiabilidade a faixa entre $P - \delta P$ e $P + \delta P$.

```python
plt.plot(1 / frequencies, power, 'k-')
plt.axvline(periodo_dominante, color='r', linestyle='--',
            label=f'Pico: {periodo_dominante:.5f} dias')
plt.axvspan(periodo_dominante - dp, periodo_dominante + dp, alpha=0.2, color='g',
            label=f'Erro: ±{dp:.5f} dias')
plt.xlabel('Período (dias)')
plt.xlim(4.21, 4.25)
plt.ylabel('Power')
plt.legend()
plt.show()
```

$\quad$ Com essa análise, observamos que o resultado é bastante confiável: o valor de período coincide com o máximo do periodograma, e o intervalo de incerteza cobre adequadamente o pico. Para validar completamente nosso resultado, comparamos nossa estimativa com os dados disponíveis na literatura. Como primeiro passo, calculamos o desvio relativo entre nossa medida e aquela reportada por Michel Mayor e Didier Queloz [3], utilizando a fórmula clássica:

$$
\Delta P = \left|\frac{P_{teo}-P_{obs}}{P_{teo}}\right| \cdot 100 \%,
$$

onde $\Delta P$ representa o desvio relativo entre os valores de período, $P_{\text{teo}}$ é o valor disponível na literatura, e $P_{\text{obs}}$ é o valor obtido neste notebook. Segundo a referência [3], o período teórico é $P_{\text{teo}} = 4{,}2293$ dias. A seguir, realizamos o cálculo do desvio.

```python
P_teo1 = 4.2293
desvio_1 = abs((P_teo1 - periodo_dominante)/P_teo1) * 100
print(f"Desvio relativo = {desvio_1:.2E} %")
```

$\quad$ Verificamos que a diferença relativa é inferior a 0,1%, o que representa um excelente nível de concordância. Em seguida, realizamos uma nova comparação, agora utilizando o valor mais recente do período orbital disponível no catálogo da NASA, o [NASA Exoplanet Archive](https://exoplanetarchive.ipac.caltech.edu/.) [13]. Ao pesquisarmos informações sobre o exoplaneta *51 Peg b*, identificamos que o artigo mais recente que caracteriza seu período orbital é o de Rosenthal et al., 2021 [14], no qual é reportado um valor de $P_{\text{teo}} = 4{,}2307969$ dias. Realizando o cálculo do desvio relativo para esse novo valor:

```python
P_teo2 =  4.2307969
desvio_2 = abs((P_teo2 - periodo_dominante)/P_teo2) * 100
print(f"Desvio relativo = {desvio_2:.2E} %")
```

$\quad$ Obtemos um desvio ainda menor! Isso é consistente com a expectativa de que medições mais recentes tendem a ser mais precisas e exatas. Sendo esse o caso, o resultado obtido neste notebook se mostra extraordinariamente preciso. Este exemplo evidencia o potencial do periodograma de Lomb-Scargle como uma ferramenta poderosa na análise de dados astronômicos.

$\quad$ Para finalizar, realizamos o procedimento conhecido como phase folding — ou seja, “dobramos” a curva temporal, reorganizando os dados dentro de um único intervalo correspondente a um período orbital. Esse processo permite analisar como os dados se distribuem ao longo de um ciclo completo. Em seguida, definimos um modelo baseado em uma função seno, que será ajustado aos dados por meio da função `curve_fit`, da biblioteca `scipy`. O modelo adotado tem a seguinte forma:

$$
R_v = A \sin{(\omega t + \phi)} + B,
$$

onde $A$ representa a amplitude, $\omega$ a frequência angular, $\phi$ a fase, e $B$ um valor de offset.

```python
# Define a função senoidal
def modelo_seno(t, A, omega, phi, B):
    return A * np.sin(omega * t + phi) + B

# Estimativas iniciais
A_inicial = (np.max(vrad) - np.min(vrad)) / 2
omega_inicial = 2 * np.pi / periodo_dominante
phi_inicial = 0
B_inicial = np.mean(vrad)
param_inicial = [A_inicial, omega_inicial, phi_inicial, B_inicial]

# Ajuste da curva
param_opt, cov = curve_fit(modelo_seno, bjd, vrad, p0=param_inicial, sigma=erro_vrad, absolute_sigma=True)
A_fit, omega_fit, phi_fit, B_fit = param_opt
```

$\quad$ Com o modelo definido, realizamos o folding da curva utilizando o período previamente determinado, restringindo a análise a um único ciclo da estrela. 

```python
# Converte tempo para fase normalizada (0 a 1)
fase = (bjd % periodo_dominante) / periodo_dominante
```

$\quad$ Também criamos um intervalo de fase compatível com as observações, permitindo a visualização simultânea dos dados medidos e do modelo ajustado em um mesmo gráfico.

```python
periodo_ajustado = 2*np.pi/omega_fit
fase_fit = np.linspace(0, 1, 1000)
t_fit_fold = fase_fit * periodo_ajustado
vrad_fit_fold = modelo_seno(t_fit_fold, *param_opt)
```

$\quad$ Por fim, exibimos o gráfico com os dados dobrados e a curva do modelo senoidal ajustado sobrepostos.

```python
plt.figure(figsize=(10, 5))
plt.errorbar(fase, vrad, yerr=erro_vrad, fmt='o', alpha=0.25, label=f'Dados dobrados (Período = {periodo_dominante:.4f} dias)')
plt.plot(fase_fit, vrad_fit_fold, color='red', label=f'Ajuste senoide (Período = {periodo_ajustado:.4f} dias)')
plt.xlabel("Fase (ciclos)")
plt.ylabel("Velocidade radial (m/s)")
plt.title(f"Curva dobrada da estrela")
plt.xlim(0, 1)
plt.legend()
plt.grid(True)
plt.show()
```

$\quad$ Nota-se que ambas as curvas apresentam um encaixe quase perfeito. Além disso, o período obtido por meio do ajuste é praticamente idêntico ao determinado pelo periodograma. Esses resultados reforçam, mais uma vez, a consistência do método aplicado e sua concordância com os dados observacionais.
