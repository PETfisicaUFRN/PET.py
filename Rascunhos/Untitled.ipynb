## 1. Introdução

Em 1857, o físico francês Jules Antoine Lissajous realizou pela primeira vez o experimento que resultaria na descoberta das figuras que iriam levar seu nome — as famosas curvas de Lissajous. Para seu experimento, Lissajous inventou um aparato que fazia com que um feixe de luz refletisse em um espelho conectado a um diapasão (pequeno instrumento metálico em forma de U, utilizado para afinar instrumentos) que vibrava em uma certa frequência, e então fosse novamente refletido por outro espelho, conectado a um segundo diapasão vibrante, orientado perpendicularmente em relação ao primeiro.

Quando o feixe de luz era refletido pelos dois espelhos conectados aos diapasões, ele desenhava intrigantes figuras na parede. A vibração de um dos diapasões fazia o feixe de luz se mover de cima para baixo enquanto a oscilação do outro o fazia se movimentar da esquerda para a direita; quando ambos vibravam ao mesmo tempo, Lissajous observou que, se os instrumentos emitissem a mesma nota, a figura tracejada seria um círculo. Experimentando com diferentes combinações de diapasões, figuras mais complexas se formavam — e Lissajous chegou à conclusão de que o único elemento que determinava a aparência destas curvas era o intervalo musical entre as notas emitidas por cada diapasão.

## 2. Notas Musicais

### 2.1 Equação da curva sinusoidal
Primeiro, começaremos discutindo o conceito mais básico da teoria musical: a existência de ***notas musicais***, sons isoláveis e distintos que formam a base de toda a música como a conhecemos. Quando um som se propaga pelo ar ou outro meio, ele o faz em forma de onda, possuindo uma *frequência* específica, e é exatamente esta frequência que irá determinar qual nota este som representa. Podemos representar as oscilações de uma nota musical usando uma curva sinusoidal, que possui um formato generalizado

$$y(t) = Asen(\omega t + \phi),$$

no qual A representa a amplitude da onda, $\omega$ é a frequência angular, e $\phi$ é a fase específica do ciclo de oscilação em $t = 0$. Considerando o fato de que a frequência angular é equivalente à frequência de oscilação da onda vezes $2\pi$,

$$\omega = 2\pi f,$$

podemos reescrever a equação da onda sinusoidal para incluir explicitamente o termo da frequência:

$$y(t) = Asen(2\pi f t + \phi).$$
