## Introdução ao mundo dos fractais

O que é um fractal, afinal? É provável que você já tenha visto alguma das seguintes imagens pela internet, em contextos relacionados à matemática ou à física. Mas muitos não sabem o que estas imagens representam, tampouco conhecem as curiosidades matemáticas por trás delas.

![mandelbrot](https://upload.wikimedia.org/wikipedia/commons/2/21/Mandel_zoom_00_mandelbrot_set.jpg)

Em sua definição geral, um fractal é uma forma geométrica que contém estruturas detalhadas a escalas arbitrariamente pequenas, geralmente com uma _dimensão fractal_ excedente à sua _dimensão topológica_ (não se preocupe, detalharei o que estes termos significam em breve). Muitos fractais apresentam semelhanças em diversas escalas, como ilustrado em ampliações sucessivas do conjunto de Mandelbrot, representado pela figura acima. Essa exibição de padrões semelhantes em escalas cada vez menores é chamada de autossimilaridade, também conhecida como simetria expansiva ou simetria de desdobramento.

Uma das maneiras pelas quais os fractais diferem de outras figuras geométricas é a forma como escalam. Dobrar o comprimento das arestas de um polígono preenchido multiplica sua área por quatro, que é dois (a razão entre o novo e o antigo comprimento da aresta) elevado à potência de dois (a dimensão convencional do polígono preenchido). Da mesma forma, se o raio de uma esfera preenchida for dobrado, seu volume aumenta por oito, que é dois (a razão entre o novo e o antigo raio) elevado à potência de três (a dimensão convencional da esfera preenchida). No entanto, se os comprimentos unidimensionais de um fractal forem todos dobrados, o conteúdo espacial do fractal escala por uma potência que não é necessariamente um número inteiro e, em geral, é maior que sua dimensão convencional. Essa potência é chamada de _dimensão fractal_ do objeto geométrico, para distingui-la da dimensão convencional, que é formalmente chamada de _dimensão topológica_.

O nome "fractal" foi escolhido pelo matemático Benoit Mandelbrot em 1975, que se baseou no latim _frāctus_, significando fratura, algo quebrado. Ele usou este termo principalmente para relacionar suas ideias teóricas sobre dimensões fractais com padrões geométricos encontrados na natureza. Estes fractais aproximados presentes no meio ambiente exibem autossimilaridade em escalas extensas, porém finitas. Alguns fenômenos naturais em que podemos encontrar características fractais são, por exemplo, vasos sanguíneos, flocos de neve, árvores, neurônios, os anéis de saturno, entre outros. Abaixo, você pode observar exemplos destes padrões na forma de cristais de gelo e "cicatrizes" causadas por raios (também chamadas de figuras de Lichtenberg).

## O floco de neve

Neste ```notebook```, iremos focar em um padrão fractal específico, conhecido como o floco de neve de Koch. Este fractal é baseado na curva de Koch, descrita pela primeira vez em 1904 pelo matemático sueco Niels Fabian Helge van Koch, em seu trabalho "Sur une courbe continue sans tangente, obtenue par une construction géométrique élémentaire" (Em uma curva contínua sem tangentes, construtível a partir da geometria elementar).

Uma curva de Koch é construída da seguinte maneira: primeiro, separamos um segmento de reta qualquer.



Em seguida, dividimos a reta em 3 partes de mesmo comprimento.



Agora, removemos a seção central e a substituímos por dois lados de um triângulo equilátero (cujos lados têm o mesmo comprimento que o segmento removido).



Agora temos uma forma composta por quatro segmentos de igual comprimento! Esta seria a primeira iteração da nossa curva de Koch. Podemos obter a segunda iteração repetindo o mesmo processo em cada um desses 4 segmentos:



Aplicando os mesmos princípios na segunda iteração, obtemos a terceira iteração.



Se continuarmos este processo infinitamente, obteremos curvas cada vez intrincadas. A curva torna-se cada vez mais longa e tortuosa, embora a distância geométrica entre os pontos finais permaneça a mesma; a linha simplesmente "se dobra" sobre si mesma.



Mas como transformamos isto em um floco de neve? Tudo que precisamos fazer é usar um triângulo equilátero na nossa primeira iteração:

![hahahakoch](https://upload.wikimedia.org/wikipedia/commons/thumb/d/d9/KochFlake.svg/960px-KochFlake.svg.png)

Uma curiosidade sobre o floco de neve Koch é que, se dermos zoom em um vértice da figura, fica fácil visualizar a autossimilaridade deste fractal. Ele sempre parece o mesmo, não importa de quão longe ou de quão perto você o observe.



No entanto, se dermos zoom em um ponto que não é vértice, a figura parece girar!



O nosso floco de neve possui uma dimensão topológica de 1, mas não é de forma alguma retificável: o comprimento da curva entre quaisquer dois pontos no floco de neve de Koch é infinito. Nenhum pequeno pedaço dele é linear, mas sim composto por um número infinito de segmentos unidos em ângulos diferentes. A dimensão fractal de uma curva pode ser explicada intuitivamente pensando em uma linha fractal como um objeto detalhado demais para ser unidimensional, mas simples demais para ser bidimensional. Portanto, sua dimensão pode ser melhor descrita não por sua dimensão topológica usual de 1, mas por sua dimensão fractal, que geralmente é um número entre um e dois; no caso do floco de neve de Koch, é aproximadamente 1,2619.
Podemos calcular isso utilizando o fator de escala ($S = 3$) e o número de objetos auto-semelhantes ($N = 4$) para calcular a dimensão, $D$, após a primeira iteração, sendo
$$D = \frac{\ln N}{\ln S} = \frac{\ln 4}{\ln 3} \approx 1,2619.$$

## Desenhando

Agora que entendemos como é formado um Floco de neve Koch, podemos desenhá-lo com ajuda do ```python```. Para isso, vamos usar uma biblioteca chamada ```turtle```, em que podemos controlar uma "tartaruguinha" que anda pela tela e desenha de acordo com o algoritmo que inserirmos no código. Somos capazes de fazer isso com o Floco de neve pois ele pertence a um grupo específico de fractais chamados _iterated function system fractals_ (fractais de sistema de função iterada), ou fractais IFS. Os principais conceitos por trás da geração desses fractais IFS são a recursão e o sistema Lindenmayer.

Se você já é familiarizada(o) com programação, deve saber que uma função recursiva é uma função que chama a si mesma durante sua execução, como a função ```koch_flake``` exemplificada abaixo. No corpo da função, utilizamos o sistema Lindenmayer (ou _L-system_), que representa um fractal como uma série de regras a ser seguidas; é um sistema de reescrita paralela e um tipo de gramática formal. Um _L-system_ consiste em um alfabeto de símbolos que podem ser usados ​​para formar cadeias de caracteres, um conjunto de regras de produção que expandem cada símbolo em uma cadeia maior de símbolos, uma cadeia inicial ("axioma") a partir da qual se inicia a construção e um mecanismo para traduzir as cadeias geradas em estruturas geométricas. Vamos observar como escreveríamos um floco de neve Koch em Lindenmayer:

axioma = “F- -F- -F”;
egrases = {“F”:”F+F- -FF+F”;”}

içõesions; =ângulo
ang0°;le = 6

Mas o que significa cada uma dessas coisas? Primeiramente, um axioma é a figura geométrica inicial. As regras são as direções em que a figura inicial deve ser alterada para se obter o fractal desejado, e o ângulo é o ângulo de rotação da nossa tartaruga. As iterações são o número de vezes que o código deve ser executado para se obter o fractal final. Já o ```F```, ```+``` e ```-``` são variáveis: ```F``` significa dar um passo à frente, ```+``` é virar à direita, e ```-``` é virar à esquerda (de acordo com o ângulo de rotação).

````python
def koch_flake(length, order): #aqui, temos uma função que recebe variáveis para tamanho e ordem do floco de neve.
    if(order == 0): #a ordem será a quantidade de vezes que teremos feito o processo de divisão de cada um dos lados do triângulo.
        forward(length)
        return
    koch_flake(length/3,order-1) #aqui, utilizamos uma função recursiva que chama a si mesma de acordo com a lógica do L-system.
    left(60)
    koch_flake(length/3, order - 1)
    right(120)
    koch_flake(length/3, order - 1)
    left(60)
    koch_flake(length/3,order-1)
````

````python
def draw_flake(): #agora, definiremos nossa função que de fato desenha nosso fractal!
    speed(0)
    up()
    backward(200 / 2.0)
    down()
    color('black')
    fillcolor('white')
    begin_fill()
    for i in range(3):
        koch_flake(300.0, 4) #chamamos a função recursiva de tamanho 300 e ordem 4
        left(-120)
    end_fill()
    mainloop()
    
draw_flake() #rodando a função
````

Este fractal obtido acima é de ordem 4. Ou seja, repetimos o processo de dividir cada linha em três e adicionar dois lados de um triângulo nela 4 vezes. Se continuarmos a aumentar a ordem, podemos observar que as bordas ficam cada vez mais tortuosas. No entanto, a área do floco de neve continua contida dentro da forma aproximada de um hexágono. Veremos a seguir que isso implica em uma propriedade bastante interessante deste fractal!

````python
def draw_flake2():
    speed(0)
    up()
    backward(200 / 2.0)
    down()
    color('black')
    fillcolor('white')
    begin_fill()
    for i in range(3):
        koch_flake(300.0, 5) #vamos colocar uma ordem maior!
        left(-120)
    end_fill()
    mainloop()
    
draw_flake2()
````

## Área vs Perímetro

Uma propriedade muito interessante do floco de neve Koch é a crescente diferença entre os valores da sua _área_ e de seu _perímetro_. Vamos analisar nosso floco para primeiro encontrar sua área: Inicialmente, o triângulo possui 3 lados, cada um com comprimento de uma unidade. Na próxima iteração, há 12 lados, cada um com comprimento de 1/3 de unidade (cada um dos três lados retos do triângulo é substituído por quatro novos segmentos). Na iteração seguinte, há 48 lados, cada um com comprimento de 1/9 de unidade (cada uma das 12 arestas anteriores é substituída por quatro novos segmentos). 
A cada nível mais profundo, criamos quatro vezes mais lados do que no nível anterio

Agora que sabemos disso, vamos ver como a área se altera a cada nova iteração. Iniciando com o triângulo equilátero, sabemos que sua área pode ser calculada de acordo com a expressão $A=\frac{\sqrt{3}}{4}s^2$.



Assim sendo, podemos concluir que o triângulo original terá a seguinte área:
$$A_0=\frac{\sqrt{3}}{4}s^2.$$

Na próxima iteração, a área do floco de neve é ​​aumentada pelos três triângulos vermelhos mostrados no diagrama abaixo 
(Três novos triângulos com lados de comprimento$s/3$3)



Assim, ficaremos com uma área total de

$$A_1=\frac{\sqrt{3}}{4}s^2+3(\frac{\sqrt{3}}{4}(\frac{s}{3})^2).$$

Na próxima iteração, adicionamos mais 12 triângulos menores. A área de cada um desses novos triângulos é somada à área total:



$$A_1=\frac{\sqrt{3}}{4}s^2+3(\frac{\sqrt{3}}{4}(\frac{s}{3})^2)+12(\frac{\sqrt{3}}{4}(\frac{s}{9})^2).$$

E assim por diante! Podemos concluir que nossa área total para $n$ iterações é dada por uma somatória.

$$A_n=\frac{\sqrt{3}s^2}{16}[4 + 3(\frac{4}{9})+ 3(\frac{4}{9})^2 + 3(\frac{4}{9})^3 + ... + 3(\frac{4}{9})^n]$$

Vamos agora dizer que a nossa série infinita se chama $G$.

$$A_n=\frac{\sqrt{3}s^2}{16}[4 + G]$$
$$G=3[\frac{4}{9}+ (\frac{4}{9})^2 + (\frac{4}{9})^3 + ... + (\frac{4}{9})^n]$$

Agora, vamos usar um truque! Sabendo $G$, conseguimos descobrir $\frac{4}{9}G$ também, apenas fazendo uma multiplicação:

$$\frac{4}{9}G=3[(\frac{4}{9})^2 + (\frac{4}{9})^3 + ... + (\frac{4}{9})^n]$$

Fazendo $G-\frac{4}{9}G$, podemos perceber que todos os termos (menos o primeiro!) se cancelam:
$$G - \frac{4}{9}G = 3(\frac{4}{9}+ (\frac{4}{9})^2 + (\frac{4}{9})^3 + ... + (\frac{4}{9})^n) - 3((\frac{4}{9})^2 + (\frac{4}{9})^3 + ... + (\frac{4}{9})^n)$$
$$\frac{5}{9}G= 3(\frac{4}{9})$$
$$G = \frac{12}{5}$$

Assim, podemos usar o valor que encontramos para $G$ para encontrar nosso valor para $A_n$:

$$A_n=\frac{\sqrt{3}s^2}{16}[4 + G]$$
$$A_n=\frac{\sqrt{3}s^2}{16}[4 + \frac{12}{5}]$$
$$A_n=\frac{\sqrt{3}s^2}{16}[\frac{32}{5}]$$
$$A_n=\frac{2\sqrt{3}s^2}{5}$$

Quando $n$ tende a infinito, é possível perceber que a área final do floco de neve será exatamente igual a $\frac{8}{5}$ da área inicial do triângulo.
:r.

Mas e quanto ao perímetro, como ele evolui com o número de iterações?

A cada passo que damos, o comprimento de cada lado é substituído por quatro conjuntos de comprimentos que correspondem a um terço do comprimento anterior. Aqui está a equação simples para o comprimento dos lados em cada profundidade:

$$L_n=3(\frac{4}{3})^n$$

Quando $n$ tende ao infinito, a fração explode e nos mostra que $L$ também tende ao infinito:

$$\lim_{n \to \infty}L_n=\lim_{n \to \infty}3(\frac{4}{3})^n=\infty$$

Ou seja, o floco de neve Koch de infinitas iterações possui uma **área finita** mas um **perímetro infinito**! Isso parece ir contra o nosso senso comum, mas mostra como fractais conseguem ser tão intrigantes e ser manter objeto de estudo de vários pesquisadores até os dias de hoje. Podemos visualizar melhor a diferença entre a área e o perímetro de acordo com o número de iterações através do gráfico abaixo:

````python
n = np.arange(0, 15)

#assumimos área inicial (A_0) = 1 e Perímetro inicial (P_0) = 3
area = (8/5) - (3/5) * (4/9)**n
perimetro = 3 * (4/3)**n

fig, ax1 = plt.subplots(figsize=(11, 6), dpi=100)

#configurando o eixo da esquerda (Área)
color_area = '#467BBD'
ax1.set_xlabel('Geração (n)', fontsize=12)
ax1.set_ylabel('Área Total', color=color_area, fontsize=12, fontweight='bold')
plot_area = ax1.plot(n, area, color=color_area, marker='o', linewidth=2.5, label='Área (Converge para 1.6)')
ax1.tick_params(axis='y', labelcolor=color_area)
ax1.set_ylim(0.8, 2.0) #escala curta para destacar a estabilização

#adicionando uma linha tracejada para mostrar o limite de convergência de 1.6
plt.axhline(y=1.6, color='#666666', linestyle='--', linewidth=1.5, label='Limite de Convergência (1.6)')

#criando o eixo da direita (Perímetro)
ax2 = ax1.twinx() 
color_perim = '#B94441'
ax2.set_ylabel('Perímetro Total', color=color_perim, fontsize=12, fontweight='bold')
plot_perim = ax2.plot(n, perimetro, color=color_perim, marker='s', linewidth=2.5, label='Perímetro (Tende ao Infinito)')
ax2.tick_params(axis='y', labelcolor=color_perim)
ax2.set_ylim(0, 100) #escala longa para o crescimento exponencial

plt.title('O Paradoxo de Koch: Área Finita vs Perímetro Infinito', fontsize=15, pad=20)
ax1.grid(True, linestyle='--', alpha=0.6)

# Combinando as legendas de ambos os eixos em uma só
lns = plot_area + plot_perim
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, loc='upper left', frameon=True, shadow=True)

fig.tight_layout()
plt.show()
````
