## 1. Introdução

$\quad$Em 1857, o físico francês Jules Antoine Lissajous realizou pela primeira vez o experimento que resultaria na descoberta das figuras que iriam levar seu nome — as famosas curvas de Lissajous. Para seu experimento, Lissajous inventou um aparato que fazia com que um feixe de luz refletisse em um espelho conectado a um diapasão (pequeno instrumento metálico em forma de U, utilizado para afinar instrumentos) que vibrava em uma certa frequência, e então fosse novamente refletido por outro espelho, conectado a um segundo diapasão vibrante, orientado perpendicularmente em relação ao primeiro.

$\quad$Quando o feixe de luz era refletido pelos dois espelhos conectados aos diapasões, ele desenhava intrigantes figuras na parede. A vibração de um dos diapasões fazia o feixe de luz se mover de cima para baixo enquanto a oscilação do outro o fazia se movimentar da esquerda para a direita; quando ambos vibravam ao mesmo tempo, Lissajous observou que, se os instrumentos emitissem a exata mesma nota, a figura tracejada seria um círculo. Substituindo um dos diapasões por um que vibrasse com o dobro da frequência, a figura tracejada era similar ao formato de uma figura oito.

$\quad$Experimentando com diferentes combinações de diapasões, figuras mais complexas se formavam — e Lissajous chegou à conclusão de que o único elemento que determinava a aparência destas curvas era o intervalo musical entre as notas emitidas por cada diapasão.

## 2. Notas Musicais

### 2.1 Equação da curva sinusoidal
$\quad$Primeiro, começaremos discutindo o conceito mais básico da teoria musical: a existência de ***notas musicais***, sons isoláveis e distintos que formam a base de toda a música como a conhecemos. Quando um som se propaga pelo ar ou outro meio, ele o faz em forma de onda, possuindo uma *frequência* específica, e é exatamente esta frequência que irá determinar qual nota este som representa. Podemos representar as oscilações de uma nota musical usando uma curva sinusoidal, que possui um formato generalizado

$$y(t) = Asen(\omega t + \phi),$$

no qual A representa a amplitude da onda, $\omega$ é a frequência angular, e $\phi$ é a fase específica do ciclo de oscilação em $t = 0$. Considerando o fato de que a frequência angular é equivalente à frequência de oscilação da onda vezes $2\pi$,

$$\omega = 2\pi f,$$

podemos reescrever a equação da onda sinusoidal para incluir explicitamente o termo da frequência:

$$y(t) = Asen(2\pi f t + \phi).$$

### 2.2 Representação de notas singulares

$\quad$Vamos primeiro olhar para a nota Dó (também conhecida como C). A nota conhecida como Dó Médio, Middle C ou C4 — a tecla exatamente no meio de um piano tradicional — é gerada por um vibração de aproximadamente $261.63 Hz$. Inserindo este valor na equação acima, considerando $A = 1$ e $\phi = 0$, obtemos o seguinte gráfico:

````python
midC = 261.63 # Dó Médio ou C4
x = arange(0, 0.01, 0.00001)
y = sin(2*pi*midC*x)

# Plotando o gráfico
plot(x, y, color='hotpink')
show()
````

$\quad$Agora, vamos analisar a nota cujo intervalo musical em relação ao Dó Médio é de exatamente uma ***oitava*** — a nota conhecida como C5, também um Dó, com uma freqûencia de aproximadamente $523.26Hz$.

````python
C5 = 523.26 # Dó 5
x = arange(0, 0.01, 0.00001)
y = sin(2*pi*C5*x)

# Plotando o gráfico
plot(x, y, color='turquoise')
show()
````

$\quad$Podemos, então, comparar ambos os gráficos.

````python
midC = 261.63 # Dó Médio ou C4
C5 = 523.26 # Dó 5 ou C5

x = arange(0, 0.01, 0.00001)
y1 = sin(2*pi*midC*x) # Curva do Dó Médio
y2 = sin(2*pi*C5*x) # Curva do Dó 5

# Plotando o gráfico
plot(x, y1, color='hotpink', label="C médio")
plot(x, y2, color='turquoise',label="C5")
legend()
show()
````

$\quad$Observando as curvas neste formato, podemos perceber que a curva da nota C5 aparenta ter uma frequência 2 vezes menor do que a frequência de C médio. Podemos visualizar melhor a relação entre as duas curvas — ou melhor, o ***intervalo musical*** entre elas — plotando suas figuras de Lissajous.

### 2.3 Curvas de Lissajous de um par de notas

$\quad$Curvas de Lissajous, também conhecidas como figuras de Lissajous ou curvas de Bowditch, são as representações gráficas de um conjunto de equações paramétricas que podem ser escritas como

$$x = Asen(at + \delta),$$

$$y = Bsen(bt).$$

$\quad$Podemos encontrar outro jeito de estudar este intervalo musical plotando e observando a curva de lissajous gerada pela oscilação conjunta de C5 e C4:

````python
A = 1 #valor da constante
B = 1 #valor da constante
passo = 0.000001
t = arange(0,0.01,passo)
tam = len(t)
````

````python
xsame = empty(tam)
ysame = empty(tam)

for i in range(tam):
    xsame[i]=A*sin(2*pi*midC*t[i]+pi/2) # Equação paramétrica para Dó Médio
    ysame[i]=B*sin(2*pi*midC*t[i]) # Equação paramétrica, também para Dó Médio

# Plotando o gráfico
plot(xsame,ysame, color='hotpink')
xlabel("middle C")
ylabel("middle C")
show()
````

$\quad$Assim como o próprio Lissajous observou em seu experimento, aqui também podemos ver que quando a razão entre duas ondulações é de $1$, a figura tracejada pela equações paramétricas (ou, no caso do cientista, pelo seu feixe de luz) é exatamente um círculo.

````python
xoctave = empty(tam)
yoctave = empty(tam)

i=0
for i in range(tam):
    xoctave[i]=A*sin(2*pi*midC*t[i]) # Equação paramétrica para Dó Médio
    yoctave[i]=B*sin(2*pi*C5*t[i]) # Equação paramétrica para C5
    i+= 1
    
# Plotando o gráfico
plot(xoctave,yoctave,color='turquoise')
xlabel("middle C")
ylabel("C5")
show()
````

$\quad$E aqui temos a figura oito encontrada por Lissajous com diapasões que vibravam em uma razão de $2:1$ um em relação ao outro.

#### 2.3.1 Intervalos musicais

$\quad$Curvas de Lissajous também são ótimas ferramentas para analisar as diferenças entre diversos intervalos musicais. Vamos começar com um dos intervalos musicais mais conhecidos, o intervalo de terça maior. Ao contrário do que muitos costumam pensar, um intervalo de terça não significa uma razão de $1:3$ entre as notas, mas sim que é o intervalo entre a primeira e a terceira nota acima desta em uma escala musical. A razão entre uma nota e sua terça é de $5:4$.

 $\quad$ Exemplo de [Intervalo de Terça](https://upload.wikimedia.org/wikipedia/commons/0/00/Third_ET.ogg) (O arquivo toca middle C seguido de E4 e, por último, ambos os tons simultaneamente).

````python
xthird = empty(tam)
ythird = empty(tam)
E4 = 327.05 # Frequência de Mi 4, em Hz

i=0
for i in range(tam):
    xthird[i]=A*sin(2*pi*midC*t[i]) # Equação paramétrica para Dó Médio
    ythird[i]=B*sin(2*pi*E4*t[i]) # Equação paramétrica para E4
    i+= 1
    
# Plotando o gráfico
plot(xthird,ythird,color='coral')
xlabel("middle C")
ylabel("E4")
show()
````

$\quad$(Se rotacionarmos a figura (adicionando um valor para $\phi$: $\phi = \pi/2$) obtemos uma visão melhor para ela neste caso específico.)

````python
for i in range(tam):
    xthird[i]=A*sin(2*pi*midC*t[i]+pi/2) # Equação paramétrica para Dó Médio, deslocada em pi/2 para melhor visualização da curva
    ythird[i]=B*sin(2*pi*E4*t[i]) # Equação paramétrica para Mi 4
    
# Plotando o gráfico  
plot(xthird,ythird,color='coral')
xlabel("middle C")
ylabel("E4")
show()
````

$\quad$O mesmo pode ser dito para intervalos de quinta, que são os intervalos entre a primeira e a quinta nota em uma escala. Os intervalos de quinta, por sua vez, possuem uma razão de $3:2$, e são conhecidos como "intervalo de quinta justa" ou, em inglês, "perfect fifth interval".

$\quad$Exemplo de [intervalo de quinta](https://commons.wikimedia.org/wiki/File:Just_perfect_fifth_on_C.mid) (o arquivo toca middle C e G4 simultaneamente, seguido de middle C, G4 e middle C novamente).

````python
xfifth = empty(tam)
yfifth = empty(tam)
G4 = 392.45 # Frequência de Sol 4, em Hz

for i in range(tam):
    xfifth[i]=A*sin(2*pi*midC*t[i]) # Equação paramétrica para Dó Médio
    yfifth[i]=B*sin(2*pi*G4*t[i]) # Equação paramétrica para Sol 4
    
# Plotando o gráfico
plot(xfifth,yfifth,color='pink')
xlabel("middle C")
ylabel("G4")
show()
````

#### 2.3.2 Dissonância

$\quad$As figuras que analisamos até agora são relativamente simples. Mas e figuras mais complexas?

$\quad$Uma curva mais complexa significa uma curva que demora mais para "fechar". Uma curva que demora mais para fechar implica em ondas que demoram mais para se encontrar, para terem um ponto em comum após o início. Por exemplo, se a razão entre as frequências de onda for um número irracional, as ondas nunca se encontrarão novamente, e a figura de lissajous resultante nunca se "completará". Já em intervalos racionais, teremos ondas que se encontram periodicamente, mesmo que percorram um longo caminho antes disso acontecer. As nossas ondas de Middle C e C5, por exemplo, se encontram periodicamente a cada oscilação completa de C5.

$\quad$Por isso, temos a impressão de que intervalos com curvas mais complexas produzem sons mais ***dissonantes***, termo utilizado para descrever a falta de harmonia entre notas. Podemos utilizar de exemplo o [intervalo de sétima menor](https://en.wikipedia.org/wiki/File:Lesser_just_minor_seventh_on_C.mid) (o arquivo tocará Middle C e B♭ simultaneamente, seguido de C, B♭ e C, respectivamente.

````python
xsev = empty(tam)
ysev = empty(tam)
Bbemol = 465.14 # Frequência de Si bemol 4

for i in range(tam):
    xsev[i]=A*cos(2*pi*midC*t[i]+pi) # Equação paramétrica para Dó Médio, deslocada em pi para melhor visualização da curva
    ysev[i]=B*cos(2*pi*Bbemol*t[i]) # Equação paramétrica para Si bemol 4
    
# Plotando o gráfico   
plot(xsev,ysev,color='violet')
xlabel("middle C")
ylabel("B♭")
show()
````

## 3. Acordes

$\quad$Um acorde é um conjunto de três ou mais notas ouvidas simultâneamente, e provavelmente são o que você visualiza quando imagina uma pessoa tocando um violão. Acordes são os blocos de construção de grande parte das músicas que escutamos hoje, e os que são formados por apenas três notas musicais distintas são denominados ***tríades***, classificados de acordo com a qualidade (maiores, menores, aumentadas e diminutas) e são os acordes que iremos analisar nesta sessão deste notebook.

$\quad$Também podemos classificar acordes com base em sua nota fundamental ou tônica, que será a nota que dá seu nome ao acorde e inicia a tríade, seguida de intervalos de terça e quinta acima de si.

$\quad$Vamos então plotar o gráfico de uma curva de Lissajous para uma tríade; consideraremos a tríade de Dó maior (você pode escutá-la [aqui](9https://upload.wikimedia.org/score/4/2/425ahbm6vzt7lnsihj4ane4qn75k4kv/425ahbm6.mp3)).

````python
fig = figure()
 
# sintaxe para projeção 3D
ax = axes(projection ='3d')

# definindo os 3 eixos
xchord = empty(tam)
ychord = empty(tam)
zchord = empty(tam)

for i in range(tam):
    zchord[i]=A*sin(2*pi*midC*t[i]) # Equação paramétrica para Dó Médio
    xchord[i]=B*sin(2*pi*E4*t[i]) # Equação paramétrica para Mi 4
    ychord[i]=B*sin(2*pi*G4*t[i]) # Equação paramétrica para Sol 4
 
# plotando o gráfico em 3D
ax.plot3D(xchord, ychord, zchord, 'lightcoral')
ax.set_title('Tríade Dó maior')
show()
````

$\quad$Algo muito interessante que podemos observar é o fato de que, se olharmos para cada uma das "faces" da figura obtida, podemos perceber que eles representam exatamente os intervalos que compõem aquele acorde específico: Terça Maior entre Middle C e E4, Quinta Justa entre Middle C e G4, e Sexta Maior entre E4 e G4. Para visualizar isso, basta plotarmos um eixo pelo outro.

````python
# Configurações dos gráficos
fig = figure(figsize=(16,6))
fig.suptitle(r"Curvas de Lissajous de cada face do acorde Dó maior", fontsize=16)

# gráfico do intervalo musical entre Mi e Sol
subplot(131)
plot(xchord,ychord,'aquamarine')

# gráfico do intervalo musical entre Mi e Dó
subplot(132)
plot(xchord,zchord,'coral')

# gráfico do intervalo musical entre Dó e Sol
subplot(133)
plot(zchord,ychord,'pink')
````
