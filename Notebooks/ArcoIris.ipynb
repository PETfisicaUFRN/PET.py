## Introdução

$\quad$ Para modelar um arco-íris, primeiro precisamos entender como a forma, tamanho e constituição de gotas líquidas afeta estes e como eles formam os arco-íris primários e secundários.

$\quad$ Quando a luz solar encontra uma gota d'água, parte da luz é refletida e o resto entra na gota. Um conjunto de gotas d'água suspensas na atmosfera funciona como um refrator de luz, já que a água representa um meio com uma densidade óptica diferente da do ar circundante, e as ondas de luz refratam quando cruzam a fronteira de um meio para outro. A diminuição da velocidade após a entrada da luz em uma gota de água causa uma curvatura do caminho da luz em direção à normal. Já ao sair da gota, a luz acelera e se desvia do normal. Ou seja, há um desvio no caminho da luz ao entrar e sair da gota.

$\quad$ O efeito geral é que parte da luz que entra é refletida de volta na faixa de 0° a 42°, com a luz mais intensa em 42°. Este ângulo é independente do tamanho da gota, mas depende do seu índice de refração. A água do mar tem um índice de refração mais alto do que a água da chuva, então o raio de um "arco-íris" na água do mar é menor do que o de um verdadeiro arco-íris. Isto é visível a olho nu por um desalinhamento desses arcos.

$\quad$ A razão pela qual a luz de retorno é mais intensa em cerca de 42° é que este é um ponto de inflexão – a luz que atinge o anel mais externo da gota é retornada a menos de 42°, assim como a luz que atinge a gota mais próxima do seu centro. Há uma faixa circular de luz que é devolvida em torno de 42°. Se o Sol fosse um laser emitindo raios paralelos e monocromáticos, então a luminância (brilho) do arco tenderia ao infinito neste ângulo se os efeitos de interferência fossem ignorados (vemos isto em Cáustica (óptica)). Mas como a luminância do Sol é finita e seus raios não são todos paralelos (ele cobre cerca de meio grau do céu), a luminância não vai ao infinito. Além disso, a quantidade pela qual a luz é refratada depende do seu comprimento de onda e, portanto, da sua cor. Este efeito é chamado de dispersão. 

$\quad$ A luz azul (comprimento de onda menor) é refratada em um ângulo maior do que a luz vermelha, mas, devido à reflexão dos raios de luz na parte de trás da gota, a luz azul emerge da gota em um ângulo menor em relação ao raio de luz branca incidente original do que a luz vermelha. Devido a esse ângulo, o azul é visto na parte interna do arco-íris primário e o vermelho na parte externa. O resultado disso não é apenas dar cores diferentes a diferentes partes do arco-íris, mas também diminuir o brilho. (Um "arco-íris" formado por gotículas de um líquido sem dispersão seria branco, mas mais brilhante do que um arco-íris normal.)

$\quad$ Existem inúmeros caminhos pelos quais os raios de luz do sol podem atravessar uma gota. Cada caminho é caracterizado por essa curvatura em direção à normal e para longe dela. Um caminho de grande importância na discussão sobre arco-íris é o caminho pelo qual a luz refrata na gota, reflete internamente e, em seguida, refrata para fora da gota.

$\quad$ Um raio de luz do sol entra na gota com uma leve trajetória descendente. Ao refratar duas vezes e refletir uma vez, o raio de luz é disperso e curvado para baixo em direção a um observador na superfície da Terra. Como no caso da refração da luz através de prismas com lados não paralelos, a refração da luz em dois contornos da gota resulta na dispersão da luz em um espectro de cores, como explicado anteriormente.

$\quad$ Tanto o arco-íris primário quanto o secundário são formados pela reflexão e refração da luz solar em pequenas gotículas de água. Quando um raio de sol é refratado duas vezes e refletido uma vez pela gotícula, um arco-íris primário se forma. Se o raio for refratado duas vezes e refletido duas vezes, um arco-íris secundário se forma. Como o arco-íris secundário é formado por uma reflexão a mais do que o arco-íris primário, ele é muito mais tênue e raro de ser visto. Por outro lado, como as trajetórias dos raios de sol em um arco-íris primário e em um arco-íris secundário são diferentes, as cores do arco-íris secundário são dispostas na ordem inversa da do arco-íris primário.

$\quad$ Abaixo, temos um exemplo de um arco-íris duplo. Perceba que as cores do arco-íris de cima, o secundário, são mais fracas e também dispostas na ordem inversa do arco-íris primário, como mencionado acima.

## Derivação Matemática

$\quad$ É possível determinar o ângulo percebido que o arco-íris subtende da seguinte maneira.

$\quad$ Vamos supor uma gota d'água esférica. Digamos que o ângulo percebido do arco-íris é de $2\varphi$, enquanto o ângulo de reflexão interna é de $2\beta$. Assim, o ângulo de incidência dos raios solares em relação à superfície da gota é de $2\beta-\varphi$. Podemos agora invocar a Lei de Snell, que nos diz que

$$ n_1\sin\theta_i=n_2\sin\theta_r.\tag{1} $$

$\quad$ Assim, utilizando os ângulos que determinamos para a gota, teremos que

$$ \sin(2\beta-\varphi)=n\sin\beta,\tag{2}$$

em que $n=\frac{n_2}{n_1}$, que neste caso será o índice refrativo da água. Resolvendo para $\varphi$, teremos

$$\varphi = 2\beta - \arcsin(n\sin\beta).\tag{3}$$

$\quad$ O arco-íris ocorre quando $\varphi$ tem um extremo em relação ao ângulo \beta, ou seja, $d\varphi/d\beta = 0$. Resolvendo para $\beta$,

$$ \beta_{max} = \arccos(\frac{2\sqrt{-1+n^2}}{\sqrt{3}n}) \approx 40.2\degree. \tag{4}$$

$\quad$ Substituindo de volta na equação para $\varphi$ e utilizando $n = 1.333$ (que é o índice refrativo da água para luz verde), vemos que $2\varphi_{max}\approx 42\degree$.

$\quad$ Algumas outras equações que serão utilizadas neste ```Notebook``` estão representadas abaixo. Além da Lei de Snell, temos também a equação do ângulo crítico, que nos diz a partir de qual ângulo de incidência não há mais arco-íris.

$$ c = \sin^{-1}(\frac{1}{\theta_r}). \tag{5}$$

$\quad$ Temos também o ângulo de desvio para o arco-íris primário, que pode ser calculado como a diferença entre o ângulo incidente e o ângulo refratado.

$$\frac{D_p}{2}=(\theta_i-\theta_r),\tag{6}$$
$$D_p=2\sin^{-1}(\frac{1-\sin\theta_i}{n}).\tag{7}$$

$\quad$ Considerando isto, o ângulo de desvio do arco-íris secundário será:

$$D_p=180-D_s.\tag{8}$$

$\quad$ Usando a Teoria Ondulatória de Airy, que nos dá uma descrição linearizada da propagação de ondas na superfície de uma camada de fluido homogêneo, concluímos que as intensidades (amplitudes) do arco-íris primário e secundário será:

$$I=(1+\cos(s-D))^2.\tag{9}$$

$\quad$ Onde $I$ é a intensidade, $s$ é o ângulo de espalhamento e $D$ o ângulo de desvio para o arco-íris individual.

## Modelagem

$\quad$ Usando as funções e equações acima, um modelo simples e um conjunto de dados foram criados e os arco-íris primários e secundários para diferentes ângulos de incidência foram plotados.

$\quad$ A função abaixo calcula as posições angulares das cores em um arco-íris primário para um determinado ângulo de incidência, índice de refração do meio (água) e raio da gota. É responsável pelo ângulo de desvio que os raios de luz sofrem após a refração e reflexão dentro da gota. A função retorna um conjunto de ângulos correspondentes a cores diferentes com base na dispersão!

````python
def calculate_primary_rainbow(angle_of_incidence, refractive_index, droplet_radius):
    # Converte ângulo para radianos
    angle_of_incidence_rad = np.radians(angle_of_incidence)
    
    # Calcula o ângulo crítico
    critical_angle = np.arcsin(1 / refractive_index)
    
    # Checa se o ângulo crítico foi excedido
    if angle_of_incidence_rad > critical_angle:
        return None  # Reflexão total interna ocorre, não há arco-íris
    
    # Calcula ângulo de desvio para o arco-íris primário
    deviation_angle = 2 * np.arcsin((1 - np.sin(angle_of_incidence_rad)) / refractive_index)
    
    # Calcula as posições angulares das cores do arco-íris
    rainbow_angles = np.degrees(np.arcsin(np.linspace(0, 1, 100) * np.sin(deviation_angle)))
    
    return rainbow_angles
````

$\quad$ Semelhante ao arco-íris primário, a função abaixo calcula as posições angulares de um arco-íris secundário. O arco-íris secundário é formado devido a duas reflexões internas dentro da gota e aparece em ângulos diferentes em relação ao primário.

````python
def calculate_secondary_rainbow(angle_of_incidence, refractive_index, droplet_radius):
    # Converte ângulo para radianos
    angle_of_incidence_rad = np.radians(angle_of_incidence)
    
    # Calcula Ângulo crítico
    critical_angle = np.arcsin(1 / refractive_index)
    
    #  Checa se o ângulo crítico foi excedido
    if angle_of_incidence_rad > critical_angle:
        return None  # Total internal reflection occurs, no rainbow
    
    # Calcula ângulo de desvio para o arco-íris primário
    deviation_angle_primary = 2 * np.arcsin((1 - np.sin(angle_of_incidence_rad)) / refractive_index)
    
    # Calcula ângulo de desvio para o arco-íris secundário
    deviation_angle_secondary = np.pi - deviation_angle_primary
    
    # Calcula as posições angulares das cores do arco-íris para o arco-íris secundário
    rainbow_angles_secondary = np.degrees(np.arcsin(np.linspace(0, 1, 100) * np.sin(deviation_angle_secondary)))
    
    return rainbow_angles_secondary
````

$\quad$ A próxima função calcula a distribuição da intensidade da luz em função dos ângulos de dispersão para arco-íris primário e secundário usando um modelo aproximado, a teoria de Airy. Ele fornece informações sobre o brilho esperado para cada parte do arco-íris.

````python
def calculate_intensity_vs_angle(angle_of_incidence, refractive_index, droplet_radius):
    #  Converte ângulo para radianos
    angle_of_incidence_rad = np.radians(angle_of_incidence)
    
    # Calcula Ângulo crítico
    critical_angle = np.arcsin(1 / refractive_index)
    
    #  Checa se o ângulo crítico foi excedido
    if angle_of_incidence_rad > critical_angle:
        return None, None, None  # Total internal reflection occurs, no rainbow
    
    # Calcula ângulo de desvio para o arco-íris primário
    deviation_angle_primary = 2 * np.arcsin((1 - np.sin(angle_of_incidence_rad)) / refractive_index)
    
    # Calcula ângulo de desvio para o arco-íris secundário
    deviation_angle_secondary = np.pi - deviation_angle_primary
    
    #Calcula ângulos de dispersão para calcular intensidade
    scattering_angles = np.linspace(0, np.pi, 100)
    
    # Calcula intensidade utilizando a teoria de Airy
    intensity_primary = (1 + np.cos(scattering_angles - deviation_angle_primary))**2
    intensity_secondary = (1 + np.cos(scattering_angles - deviation_angle_secondary))**2
    
    return np.degrees(scattering_angles), intensity_primary, intensity_secondary
````

$\quad$ O código traça as posições angulares das cores dos arco-íris primários e secundários em função do ângulo de incidência, e usa um loop para calcular e traçar arco-íris para diferentes ângulos de incidência. Cada tipo de arco-íris é representado por um estilo de linha diferente no gráfico.  

$\quad$ O modelo percorre uma faixa de ângulos de incidência de 0 a 90 graus e calcula as posições dos arco-íris primários e secundários para cada ângulo.

````python
# Parâmetros
refractive_index_water = 1.333  # Refractive index of water

# Alcance do ângulo incidente(em graus)
incident_angles = np.linspace(0, 90, 91)

# Tamanho da gotícula em micrômetros
droplet_size = 1000.0  # Pode ser ajustado para testar tamanhos diferentes!

# Plotando arco-íris primários e secundários para diferentes ângulos de incidência
plt.figure(figsize=(10, 6))

for angle in incident_angles:
    rainbow_angles = calculate_primary_rainbow(angle, refractive_index_water, droplet_size)
    if rainbow_angles is not None:
        plt.plot(rainbow_angles, np.linspace(0, 1, len(rainbow_angles)), label=f'Arco-íris Primário (Ângulo Incidente: {angle}°)')
    
    rainbow_angles_secondary = calculate_secondary_rainbow(angle, refractive_index_water, droplet_size)
    if rainbow_angles_secondary is not None:
        plt.plot(rainbow_angles_secondary, np.linspace(0, 1, len(rainbow_angles_secondary)), linestyle='--', label=f'Arco-íris Secundário (Ângulo Incidente: {angle}°)')

plt.xlabel('Ângulo (graus)')
plt.ylabel('Intensidade Normalizada')
plt.title(f'Arco-íris Primário e Secundário para Gotícula do Tamanho: {droplet_size} μm')
plt.xlim(0, 180)
plt.ylim(0, 1)
plt.legend()
plt.grid(True)
plt.show()
````

$\quad$ Depois de examinar as posições angulares, o código calcula e traça as distribuições de intensidade para um ângulo de incidência específico usando a função ```calculate_intensity_vs_angle```. Isto fornece uma representação visual de como a intensidade da luz varia em diferentes ângulos de dispersão para os arco-íris primário e secundário. 

$\quad$ O gráfico abaixo exibe os padrões de intensidade para um ângulo de incidência específico ($40$ graus neste exemplo) para o arco-íris primário e secundário.

````python
# Calcular a intensidade vs. ângulo de espalhamento para um ângulo incidente específico
angle_of_incidence_example = 40  # Exemplo de ângulo incidente em graus
scattering_angles, intensity_primary_example, intensity_secondary_example = calculate_intensity_vs_angle(angle_of_incidence_example, refractive_index_water, droplet_size)

# Plotando intensidade vs. ângulo de espalhamento
plt.figure(figsize=(10, 6))
plt.plot(scattering_angles, intensity_primary_example, color='blue', label='Arco-íris Primário')
plt.plot(scattering_angles, intensity_secondary_example, color='red', linestyle='--', label='Arco-íris Secundário')
plt.xlabel('Ângulo de Espalhamento (graus)')
plt.ylabel('Intensidade')
plt.title("Intensidade vs. Ângulo de Espalhamento")
plt.legend()
plt.grid(True)
plt.show()
````
