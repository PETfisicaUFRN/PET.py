## Posição das estrelas

Para podermos estudar as estrelas, primeiro é necessário entender as posições do céu. Saber as coordenadas não implica em saber a natureza da estrela e nem suas propriedades físico-químicas. No entanto, as coordenadas se tornam úteis quando precisamos analisar uma estrela de maneira individual ou um grupo de estrelas próximas. O leitor em algum momento deve ter ouvido falar das constelações, que de fato são uma boa maneira de organizar cada região do céu, mas nós precisamos de algo mais geral para descrever as posições. Podemos descrever a posição das estrelas por meio de duas quantidades. Desde de que nós usamos um sistema de coordenada que é fixa em relação a Terra. A Terra tem rotação, mas nós não queremos um sistema de coordenadas que rotacione, pois causaria muitos problemas. Portanto, usamos coordenadas fixadas no espaço. O plano equatorial de nossa esfera celeste é o mesmo que o plano equatorial da Terra, o que por sua vez quer dizer que o plano equatorial é perpendicular à rotação da Terra. No entanto, o eixo de rotação não é fixo, pois devido à atração gravitacional do Sol e da Lua, a Terra sofre com o movimento da precessão, que se trata de um fenômeno físico que consiste na mudança do eixo de rotação de um objeto, causando um efeito giroscópico observado nos movimentos dos pontos de referência celestes. Por causa disso, a posição das estrelas do céu celeste mudam. Obviamente as estrelas também têm seu movimento próprio, mas a alteração da posição das estrelas é liderada pela ação da precessão.

As coordenadas que os astrônomos usam são a ascensão reta $\alpha$ e a declinação $\delta$. A ascensão reta corresponde à longitude, que nós usamos para descrever a posição de um lugar na Terra e a declinação corresponde a latitude que nós usamos na superfície da Terra. Como nós já sabemos, nós definimos na Terra um meridiano que serve como referência, chamamos de longitude zero. Na Terra chamamos esse meridiano de meridiano de Greenwich. Na esfera celeste, também definimos uma posição com base na longitude ou ascensão reta. Poderíamos definir a posição de uma dada estrela como posição zero, porém a estrela pode se mover e o sistema de coordenada iria se alterar de acordo com o movimento da estrela escolhida. Então, para resolver essa incógnita, foi definido como ponto zero para a ascensão reta, a interseção de dois planos, o plano equatorial e a eclíptica, é a projeção sobre a esfera celeste da trajetória aparente do Sol observada a partir da Terra. A orientação do plano equatorial também muda com o tempo, devido a precessão do eixo de rotação da Terra,o
que por sua vez causa a mudança da intersecção do plano equatorial e a eclíptica. Ou seja, a coordenada zero da ascensão reta muda com o tempo. Então, as coordenadas para todas as estrelas mudam com o tempo, mas que podem ser computadas da movimentação do eixo de rotação da Terra. A ascensão reta é medida em horas ou em graus, indo de 0 à 24h ou de 0º à 360º,respectivamente. A declinação é medida em graus e tem como origem o plano equatorial, e tendo como intervalo -90º < $\delta$ < + 90º. 

## Movimento próprio das estrelas

As coordenadas das estrelas mudam majoritariamente devido ao movimento de precessão da Terra. Todavia, as coordenadas das estrelas podem também ser mudadas devido ao movimento que elas mesmas fazem no espaço. Porém, apenas movimentos perpendiculares à linha de visada podem alterar a posição da estrela na esfera celeste. Pois, movimentos da estrela na linha de visada podem alterar a distância delas em relação a Terra, mas não a posição dela na esfera celeste. Esses movimentos perpendiculares à linha de visada são denominados de movimento próprio, pois como o próprio nome sugere, é o movimento causado pela própria estrela no espaço. O movimento próprio é mensurado em sec ou arcsec por ano, em contrapartida a velocidade radial é medida em $km s^{-1}$. Não é conveniente determinar a velocidade do movimento próprio em $\frac{Km}{s}$ porque essa velocidade depende da distância da estrela. A distância de uma estrela, a uma dada velocidade, vai determinar o quanto ela vai se deslocar na esfera celeste, por exemplo, uma estrela mais próxima tem um movimento próprio em arcsec maior do que uma estrela mais afastada. 

Obs.: Uma estrela tem a velocidade radial, que se trata da velocidade da estrela na nossa linha de visada, a velocidade transversal que 
influência no movimento próprio, e a velocidade espacial que se trata do movimento resultante da estrela.

## Distancia de estrela próximas

A distância de objetos próximos na Terra são determinados em unidades de metro, medindo usando uma régua. Porém, para medir distâncias muito maiores como por exemplo a distância do topo de duas montanhas. 

Por exemplo, nós determinamos a distância usando um método chamado de triangulação. Nós observamos um objeto de diferentes pontos de vista e medimos a distância. Dois pontos $A$ e $B$ observam um objeto no ponto $C$ que vai aparecer projetado contra o fundo em diferentes posições $E$ e $D$. Para objetos próximos vamos usar um ângulo $\gamma$ entre a projeção dos pontos, quanto maior a distância do alvo menor este ângulo. A relação do ângulo $\gamma$ com a distância é dada por:

$$\sin\left(\frac{\gamma}{2}\right) = \frac{a}{2d}$$

$$\sin\left(\frac{\gamma}{2}\right) = \frac{a}{2d} \tag{1,1}$$

Sendo $a$ a distância entre os dois pontos de observação A e B, e d é a distância do objeto.

Para distâncias muito grandes nós podemos igualar $\sin(\gamma) = \gamma$, podemos então reescrever a fórmula como sendo: 

$$\gamma = \frac{a}{d} \tag{2,2}$$

Para nós, o $a$ serve como a distância entre nossos olhos. 

Tecnicamente falando, nós podemos medir distância muito longas usando um $a$ muito grande, porém a Terra tem um limite na distância $a$ em que o ângulo $\gamma$ pode ser determinado. A Terra tem como limite para $a$ o seu diâmetro. Não é o bastante para se medir a distância da Terra-Sol, mas podemos determinar a distância de asteroides próximos e usar a terceira lei de Kepler para determinar a distância do Sol. Usando a relação:

$$ \frac{P^2}{b^3} = const = A \tag{3,3}$$

Então, para a distância da Terra e Vênus( ou um asteroide):

$$ \frac{P(Venus)^2}{P(Terra)^2} = \frac{b(Venus)^3}{b(Terra)^3} \tag{4,4}$$

Onde $b(Venus)$ e $b(Terra)$ se tratam do semi-eixo maior da órbita de Vênus e Terra. $P$ é o período, sendo $P(Venus)$ o período de Vênus e $P(Terra)$ período da Terra(1 ano).

Aproximando a órbita dos planetas em círculos, em vez de elípticas, podemos determinar a distância de Vênus e Terra na sua máxima aproximação um em relação ao outro, usando a seguinte relação:

$$ d = b(Vênus) - b(Terra) \tag{5,5} $$

A distância Terra-Sol é 1 au(unidade astronômica), a distância Vênus-Sol é de apenas 0.72 au e a distância Terra-Vênus mais próxima é de 0.28au.

Nós escrevemos a distancia Terra-Sol como sendo:

$$ 1 au = 1.49 x 10^{13} cm$$

 ### Trigonometria da paralaxe das estrelas 

Observando uma  estrela em 1 dia e após meio ano nós observamos novamente, porém devido ao movimento da translação estaremos observando a mesma estrela em uma posição diferente, possibilitando assim a medida da paralaxe. A mudança de posição em meio ano é dada por $ângulo (2\pi) = 2au/d$ ou simplesmente.

$$\pi = \frac{1au}{d} \tag{6,6}$$

Onde a letra d indica a distância da Terra em relação ao Sol. Agora, sabendo a distância em au(unidade astronômica) e medindo $\pi$ vai nos dá a distância da estrela. Logo, desde que as distâncias das estrelas são muito grandes é mais conveniente introduzir a uma nova unidade de medida para comprimento. Nós chamamos de parsec que é a distância de uma estrela que tem como ângulo paralaxe $\pi$ igual a 1 arcsec. Nós podemos mensurar ângulos paralaxe maior do que 0.05 arcsec.

## Calculando a distância 

No código abaixo iremos determinar a distância do aglomerado Hyades utilizando o método da paralaxe e as ferramentas oferecidas pela linguagem de programação python. 

### **Importe as bibliotecas**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import astropy.table
```

Após baixar os dados, utilize as funções abaixo para ler o arquivo vot e converter para pandas, e assim fazer um dataframe.

```python
table = astropy.table.Table.read('Cone_gaiadr3.gaia_source_[67.447,16.948]_4.0.vot',format='votable')
df = table.to_pandas()
df.head()
```

No dataframe acima, usaremos as colunas de ra(ascensão reta), dec(declinação), parallax, pmra( movimento próprio na ascensão reta), pmdec(movimento prórpio na declinação) e g_rp(magnitude aparente). Portanto, para melhor visualização, podemos apenas utilizar as colunas necessárias para o presente notebook e fazer um novo dataframe.

```python
Colunas_usadas = df[['ra','dec','parallax','pmra','pmdec','g_rp']]
Colunas_usadas.head(10)
```

Abaixo, vemos que a quantidade de objetos que estamos analisando é de 463802.

```python
len(df)
```

Utilizando a ascensão reta e a declinação, fazemos uma projeção Mollweide e visualizamos a grande quantidade de objetos analisados. 

```python
plt.style.use('dark_background')
plt.rc('font',size=15)
```

```python
plt.scatter(df['ra'],df['dec'],s=1,alpha=0.03,color='red')
plt.xlabel('RA')
plt.ylabel('DEC')
plt.show()
```

Sabendo que o aglomerado  Hyades é visível a olho nú, podemos colocar uma condição que permita analisarmos apenas objetos que sejam úteis. Nesse caso, buscamos a coluna com nome "g_rp", que se trata dos dados das magnitudes aparentes da estrela desse aglomerado, colocamos a magnitude aparente menor do que 1.2.

```python
df_r = df.loc[df['g_rp']<1.2]
len(df_r)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(df_r['pmra'],df_r['pmdec'],s=0.05,alpha=1,color='red')
plt.xlabel('PMRA')
plt.ylabel('PMDEC')
plt.show()
```

O gráfico acima representa o movimento próprio da estrela na declinação e na ascensão reta, onde a maior parte dos pontos são dispersões, erros de medidas e outros dados de outros objetos celestes.

Entretanto, ao ampliarmos a imagem, observamos um "detalhe/caroço" que se destaca como o nosso aglomerado. Onde é possível notar um acúmulo de objetos celestes que não estão no centro do gráfico.

```python
plt.figure(figsize = (10,8))
plt.scatter(df_r['pmra'],df_r['pmdec'], s = 5, alpha = 0.1, color='red')

plt.xlabel('pmra')
plt.ylabel('pmdec')

plt.xlim(-200,200)
plt.ylim(-200,200)
plt.show()
```

Conhecendo os valores do pmra e pmdec, e considerando que, por ser um aglomerado, o movimento das estrelas será semelhante e elas estarão mais próximas entre si. Portanto, é suficiente analisar as posições no gráfico e selecionar as posições do gráfico em que a um acumulo de objetos celeste fora do centro de pontos.

```python
# a medida de cada eixo está em mili-segundo de arco por ano em cada eixo

comoving = df.loc[df['pmra'] <= 150]
comoving = comoving.loc[comoving['pmra'] >= 90]
comoving = comoving.loc[comoving['pmdec'] <= 0]
comoving = comoving.loc[comoving['pmdec'] >= -50]
```

Uma vez encontradas as estrelas do aglomerado Hyades, podemos fazer um histograma que mostre a o intervalo da paralaxe das estrelas desse aglomerado.

```python
plt.figure(figsize=(10,8))
plt.hist(comoving['parallax'],bins = 80, color ='blue',alpha = 0.8)
plt.xlabel('parallax')
plt.ylabel('Números de estrelas')
```

```python
#paralax em milisegundos de arco
hyades = comoving.loc[comoving['parallax'] <= 25]
hyades = hyades.loc[comoving['parallax'] >= 17]
```

Utilizando as barras com o intervalo do histograma com o maior número de estrelas, podemos encontrar a paralaxe do aglomerado.

```python
#Pegando os dados  da paralaxe das estrelas do aglomerado Hyades, obtidos pelo satélite Gaia e selecionados pelo código
#temos que:
p = np.mean(hyades['parallax'])
print(p) # paralaxe
```

```python
centauri_dist_parsec = (1000/p) #Distância em Parsec
print(centauri_dist_parsec)
```

**Obs.: 1 parsec equivale a 3.26 ano-luz**

```python
print(centauri_dist_parsec*3.26) #Distância em ano-luz
```

Sabemos que a paralaxe do aglomerado Hyades é de cerca de 22 arsec e que a distância é $\pm$ 151 ano-luz. Deste modo conseguimos determinar a distância do aglomerado Hyades com uma pequena margem de erro.

Ao aplicar os conceitos trigonométricos discutidos neste notebook, calculamos a distância do aglomerado Hyades com base nas medidas de paralaxe obtidas. A precisão dessa técnica é notável, fornecendo uma compreensão mais profunda das distâncias interestelares.

O aglomerado Hyades, por sua proximidade e disposição de estrelas, serviu como um excelente exemplo prático. Observamos como a técnica de paralaxe é especialmente eficaz quando aplicada a aglomerados, onde as estrelas compartilham movimentos semelhantes e estão mais concentradas.

Ao analisar graficamente as posições das estrelas do aglomerado no decorrer do tempo, notamos uma clara tendência, facilitando a identificação das estrelas centrais ou do "caroço" do aglomerado.
