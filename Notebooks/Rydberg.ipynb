# 1. Introdução

$\quad$ Nesta seção daremos a introdução teórica necessária para entender o que são átomos de Rydberg, como eles interagem entre si e com lasers numa rede e desenvolver o formalismo de estados produto (campo médio) para eles.

## 1.1. Átomos de Rydberg

$\quad$ Átomos de Rydberg são átomos em que um elétron da camada de valência está em um estado muito excitado (número quântico principal alto). Atualmente, átomos de Rydberg são um grande foco de pesquisa na Física Atômica devido ao fato de que, devido às escalas de energia e propriedades exageradas destes objetos, é possível realizar com eles experimentos que seriam impossíveis com outros tipos de átomo [1].

$\quad$ Dado que tenhamos vários átomos em uma rede periódica com separação muito maior do que o tamanho de cada átomo no estado fundamental, presos por algum potencial qualquer e individualmente sob a ação de lasers que excitam os átomos para estados de Rydberg, podemos tratar cada átomo como um sistema de dois níveis. Sejam $|r\rangle = (1, 0)^T$ e $|g\rangle = (0, 1)^T$ vetores de estado representando um átomo no estado excitado (Rydberg) e no estado fundamental (ground-state), respectivamente, podemos definir os projetores $|r\rangle \langle r| = \begin{pmatrix} 1 & 0 \\ 0 & 0 \end{pmatrix}$ e $|g\rangle \langle g| = \begin{pmatrix} 0 & 0 \\ 0 & 1 \end{pmatrix}$ que agem nos kets $|r\rangle$ e $|g\rangle$ da seguinte maneira:

$$|r\rangle \langle r| |r\rangle = |r\rangle, |r\rangle \langle r||g\rangle = 0, \tag{1}$$


$$|g\rangle \langle g||r\rangle = 0, |g\rangle \langle g||g\rangle = |g\rangle, \tag{2}$$

ou seja, $|r\rangle \langle r|$ é o operador que nos diz o número de átomos excitados, enquanto $|g\rangle \langle g|$ faz o mesmo para o ground-state.

$\quad$ Podemos definir também os operadores $|r\rangle \langle g| = \begin{pmatrix} 0 & 1 \\ 0 & 0 \end{pmatrix}$ e $|g\rangle \langle r| = \begin{pmatrix} 0 & 0 \\ 1 & 0 \end{pmatrix}$, tais que

$$|r\rangle \langle g| |r\rangle = 0, |r\rangle \langle g||g\rangle = |r\rangle, \tag{3}$$

$$|g\rangle \langle r||r\rangle = |g\rangle, |g\rangle \langle r||g\rangle = 0, \tag{4}$$

ou seja, estes operadores transformam estados de Rydberg em estados fundamentais e vice-versa, correspondendo assim a operadores de transições atômicas [2, 3, 4].

$\quad$ A partir destes operadores, podemos construir os primeiros dois termos da matriz Hamiltoniana do nosso sistema. Chamemos de $n_a$ o operador $|r\rangle \langle r|$ agindo no sítio $a$ da nossa rede e notemos que ele é exatamente igual a $\frac{1 + \sigma^z}{2}$, enquanto $|r\rangle \langle g| + |g\rangle \langle r|$ é exatamente a matriz de Pauli $\sigma^x$, onde as matrizes de Pauli são $\sigma^x =\begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix}$, $\sigma^y =\begin{pmatrix} 0 & -i \\ i & 0 \end{pmatrix}$ e $\sigma^z = \begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix}$. Podemos então escrever a Hamiltoniana como sendo [3, 4]:

$$H = H_{int} + \frac{\Omega}{2}\sum_a \sigma_a^x - \Delta \sum_a n_a, \tag{5}$$

onde $H_{int}$ é a componente de interação interatômica, $\Omega$ é a frequência de Rabi do laser que excita os átomos e $\Delta$ é o detuning do laser, que mede a diferença entre a frequência do laser e a frequência natural de transição dos átomos. Notemos que, na equação acima e para o resto do `Notebook`, utilizamos um sistema de unidades onde a constante de Planck reduzida é $\hbar = 1$. Para determinar a forma do termo de interação na Hamiltoniana, notemos que, como os átomos estão bastante separados e são neutros, podemos desconsiderar interações entre átomos neutros. Quando os átomos estão em estados de Rydberg, por outro lado, cada um tem um elétron com número quântico principal muito alto, ou seja, um elétron que tende a estar muito distante do núcleo. Na prática, são apenas esses elétrons que interagem entre si repulsivamente, gerando uma contribuição positiva somente entre átomos que estão ambos excitados. Assim, temos [3, 4]:

$$H = \sum_{a < b} V_{ab} n_a n_b + \frac{\Omega}{2}\sum_a \sigma_a^x - \Delta \sum_a n_a. \tag{6}$$

$\quad$ Na expressão acima, $V_{ab} = \frac{C_6}{r_{ab}^6}$ é uma interação de Van der Waals, onde $C_6$ é uma constante e $r_{ab}$ é a distância entre os átomos $a$ e $b$. A interação de Van der Waals na rede tem uma propriedade interessante: como a contribuição de energia é maior quanto mais átomos estiverem excitados, o sistema tende a tentar reduzir a interação para reduzir a sua energia: quando um átmomo está excitado, existe um raio de bloqueio $R_b$ dentro do qual nenhum outro átomo pode estar excitado em um dado momento. O raio de bloqueio $R_b$ está relacionado com os outros parâmetros do sistema através da seguinte expressão [3, 4]:

$$C_6 = \Omega\left(\frac{R_b}{a}\right)^6. \tag{7}$$

$\quad$ Embora a interação de Van der Waals tente diminuir o número de átomos excitados, o termo de detuning $\Delta$ contribui para aumentar esse número. Assim, vemos que nosso sistema tem interações diferentes que competem conforme alteramos os parâmetros $R_b$, $\Omega$ e $\Delta$ [3, 4].

## 1.2. Aproximação por Estados Produto

$\quad$ Para trabalhar com o nosso sistema de átomos de Rydberg na prática, ao longo deste `Notebook`, utilizaremos uma aproximação de campo médio, em que trataremos cada átomo como um sistema independente que interage com os outros. Ou seja, desconsideraremos a possibilidade de emaranhamento quântico entre os átomos e presumiremos que, para um estado $|\Psi\rangle$ qualquer, o valor esperado $ \langle n_1 n_2 \rangle = \langle \Psi| n_1 n_2 |\Psi\rangle$ pode ser decomposto na forma

$$\langle n_1 n_2 \rangle = \langle n_1\rangle \langle n_2 \rangle. \tag{8}$$

$\quad$ Como estamos tratando cada átomo separadamente, podemos parametrizar o estado de cada átomo na forma canônica para um sistema de dois níveis:

$$|\Psi_a\rangle = \cos(\theta_a)|r\rangle + e^{i\phi_a}\sin(\theta_a)|g\rangle. \tag{9}$$

Os $\theta_a$ e os $\phi_a$ na expressão acima são, no geral, diferentes para cada átomo da rede. Com base nesta parametrização, podemos escrever o valor esperado das matrizes de Pauli ($s^x_a = \left< \hat{\sigma}_a^x \right>$, e analogamente para $s^y_a$ e $s^z_a$) como sendo:

$$s_a^x = \cos(\varphi_a) \sin(2 \theta_a), \tag{10}$$

$$s_a^y = \sin(\varphi_a) \sin(2 \theta_a), \tag{11}$$

$$s_a^z = \cos^2(\theta_a) \sin^2(\theta_a) \tag{12}.$$

$\quad$ O valor esperado da energia também pode ser escrito de forma semelhante:

$$\left< E_{PS} \right> =  \sum_{a<b} V_{ab} \left< \hat{n}_a \right> \left< \hat{n}_b \right> + \frac{\Omega}{2} \sum_a \left< \hat{\sigma}_a^x \right> - \Delta \sum_a \left< \hat{n}_a \right>. \tag{13}$$

$\quad$ Utilizando $n_a = \frac{1 + \sigma_a^z}{2}$ junto às equações (10) e (12), temos:

$$\left< E_{PS} \right> =  \sum_{a<b} V_{ab} \frac{(s_a^z + 1 )}{2} \frac{(s_b^z +1)}{2} + \frac{\Omega}{2} \sum_a s_a^x - \Delta \sum_a \frac{(s_a^z +1)}{2}, \tag{14}$$

$$\left< E_{PS} \right> =  \sum_{a<b} V_{ab} \cos^2(\theta_a) \cos^2(\theta_b) + \frac{\Omega}{2} \sum_a \cos(\varphi_a) \sin(2 \theta_a) - \Delta \sum_a \cos^2(\theta_a). \tag{15}$$

$\quad$ Temos na equação (15) uma expressão para a energia do sistema em função dos $2N$ (onde $N$ é o número de átomos na rede) parâmetros ${\theta}$ e ${\phi}$ Com isso, podemos empregar o princípio variacional e minimizar a energia em função destes parâmetros para achar um limite superior para a verdadeira energia do ground-state, assim aproximando-a.

# 2. Minimização da Energia

$\quad$ Para começar nosso trabalho de otimização, primeiro precisamos construir numericamente a nossa rede atômica. Definamos para isso a função `rede`, que constrói uma rede quadrada com $N_{lado}$ átomos de cada lado, com átomos vizinhos separados por uma distância igual ao parâmetro de rede $a$:

```python
# Dado o número de átomos em um lado do quadrado e a distância entre átomos vizinhos, retorna os vetores posição da rede
def rede(N, a):
    vecs = [] # Inicializa o conjunto dos vetores
    for i in range(N):
        for j in range(N):
            vecs.append(jnp.array([i*a, j*a])) # Adiciona um vetor da rede a vecs
    return jnp.array(vecs)
```

$\quad$ Tomemos nossa rede como sendo um quadrado de $3$ átomos de lado e parâmetro de rede $a = 1$.

```python
lado = 3 # Número de átomos no lado do quadrado
N = lado**2 # Número total de átomos na rede
a = 1 # Parâmetro de rede (distância entre átomos vizinhos)
```

$\quad$ Podemos agora utilizar um _scatter plot_ para visualizar nossa rede no espaço:

```python
# Plotando a rede
vecs_rede = rede(lado, a)
plt.scatter(vecs_rede[:, 0], vecs_rede[:, 1])
```

$\quad$ Tendo visto que nossa função `rede` realmente constrói uma rede quadrada, podemos usas as posições da rede calculadas por ela para computar o valor das interações interatômicas, que fazemos através da função `V_int`:

```python
# Construindo as interações de Van der Waals entre os átomos da rede
def V_int(C6, r):
    Vij = np.zeros((N,N)) # Inicializando a matriz de interação
    for i in range(N):
        for j in range(N):
            rij = np.linalg.norm(vecs_rede[i]-vecs_rede[j]) # Distância entre os átomos i e j
            if rij > 0: # Evitando divisão por 0 (sem interações de um átomo consigo mesmo)
                Vij[i][j] = C6/rij**6 # Adicionando o valor da energia potencial de interação à matriz Vij
    return Vij
```

$\quad$ Dada a matriz de interação construída, podemos, tendo valores de $\Delta$, dos  $V_{ij}$, $\Omega$ e um chute inicial para os nossos conjuntos de parâmetros, computar a energia conforme a equação (15). Para isso, definimos a função `var_en` e utilizamos o _wrapper_ do `jax` em conjunto com a função `value_and_grad`, da mesma biblioteca, para definir uma função que computa de forma extremamente eficiente o valor e o gradiente (em função dos parâmetros) da energia do sistema.

```python
def var_en(param, Vij, Om, Delt):
    
    theta = param[:N] # Inicializando os parâmetros theta
    phi = param[N:] # Inicializando os parâmetros phi
    
    cos2 = jnp.cos(theta)**2 # cos^2 de theta
    Cosij = jnp.outer(cos2, cos2) # matriz com os resultados acima
    
    en = 0.5*jnp.sum(Vij*Cosij) + jnp.sum(0.5*Om*jnp.cos(phi)*jnp.sin(2*theta) - Delt*cos2) # somando as contribuições de energia
    
    return en

PS_fdf = jit(value_and_grad(var_en)) # definindo o wrapper just-in-time do jax para o valor e o gradiente da função var_en para a otimização
```

$\quad$ Com isso tudo pronto, podemos tentar encontrar o ground-state do nosso sistema. Definamos valores para as constantes de acoplamento e calculemos as intensidades das interações correspondentes:

```python
Delta = 1.5 # Valor de Delta
Omega = 0.0 # Valor de Omega
C6 = 1.0 # Valor de C6
```

```python
Vij = V_int(C6, vecs_rede) # Calculando as interações
```

$\quad$ Para tentar encontrar o estado fundamental, utilizaremos 50 iterações do método de Basinhopping, criado para imitar o comportamento de átomos sendo resfriados, através da função `basinhopping` do módulo `scipy.optimize`:

```python
n_iter = 50 # Número de iterações do otimizador basinhopping

minimizer_kwargs = {"args":(Vij, Omega, Delta), "method":"L-BFGS-B", "jac":True} # Parâmetros adicionais a passar para o otimizador

x0 = 2*np.pi*np.random.random(2*N) # Inicializando os parâmetros theta e phi em uma configuração aleatória

result_PS = basinhopping(PS_fdf, x0, minimizer_kwargs=minimizer_kwargs, niter=n_iter) # Realizando a otimização

E_PS =  result_PS.fun # Energia minimizada
state_PS = result_PS.x # Parâmetros theta e phi que minimizam a energia
```

```python
print(f'Energia do Ground-State: {E_PS}')
```

$\quad$ Temos nossa primeira estimativa para a energia! Para verificar como o estado definido pelos parâmetros otimizados se parece na rede, computaremos $n_z = \frac{1 + \sigma^z}{2}$ para cada sítio e analogamente para as outras matrizes de Pauli:

```python
nz = np.cos(state_PS[:N])**2 # Calculando o número de excitações
nx = (1 + np.sin(2*state_PS[:N])*np.cos(state_PS[N:]))/2 # Analogamente para sigmax
ny = (1 + np.sin(2*state_PS[:N])*np.sin(state_PS[N:]))/2 # Analogamente para sigmay
```

```python
# Plotando os números de excitação em cada ponto da rede
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, sharey=False, sharex=False, figsize=(15,5))

pc1 = ax1.scatter(vecs_rede[:,0], vecs_rede[:,1], c= nz, vmin=0., vmax=1.0, cmap=cm.bwr, s=100)

pc2 = ax2.scatter(vecs_rede[:,0], vecs_rede[:,1], c= nx, vmin=0., vmax=1.0, cmap=cm.summer, s=100)

pc3 = ax3.scatter(vecs_rede[:,0], vecs_rede[:,1], c= ny, vmin=0., vmax=1.0, cmap=cm.summer, s=100)

cb1 = fig.colorbar(pc1, ax=ax1, location='right')
cb2 = fig.colorbar(pc2, ax=ax2)
cb3 = fig.colorbar(pc3, ax=ax3)


ax1.axes.get_xaxis().set_visible(False)
ax1.axes.get_yaxis().set_visible(False)

ax2.axes.get_xaxis().set_visible(False)
ax2.axes.get_yaxis().set_visible(False)

ax3.axes.get_xaxis().set_visible(False)
ax3.axes.get_yaxis().set_visible(False)

ax1.set_title(r'$n_z$')
ax2.set_title(r'$n_x$')
ax3.set_title(r'$n_{y}$')

plt.tight_layout()
plt.show()
plt.close()
```

$\quad$ Notamos que, para $n_x$ e $n_y$, todos os átomos estão constantes em $0.5$. Isso não é nenhuma surpresa, já que definimos $\Omega = 0$ tal que os únicos termos que sobrevivem no Hamiltoniano dependem apenas de $\sigma^z$. Para $n_z$, por outro lado, todos os átomos estão ou em $n_z = 1$ ou em $n_z = 0$, ou seja, estão excitados ou não-excitados. Porém, algo parece estranho: há átomos vizinhos excitados e não parece haver nenhuma ordem no sistema. Será que este é realmente o estado fundamental ou nossa otimização não convergiu direito? Vamos tentar otimizar várias vezes para verificar isso:

```python
# Minimizando a energia 10 vezes para testar se o número de iterações é alto o suficiente

for i in range(10):
    x0 = 2*np.pi*np.random.random(2*N)
    
    result_PS = basinhopping(PS_fdf, x0, minimizer_kwargs=minimizer_kwargs, niter=n_iter)
    
    E_PS =  result_PS.fun
    print(E_PS)
```

$\quad$ Comparando os valores acima com a energia que havíamos encontrado para nosso ground-state, vemos que não era a menor possível! Isso indica que 50 iterações de _basinhopping_ não são suficientes. Aumentemos o número de iterações para 500, de forma a garantir uma convergência melhor.

```python
n_iter = 500 # Número de iterações atualizado

minimizer_kwargs = {"args":(Vij, Omega, Delta), "method":"L-BFGS-B", "jac":True} # Parâmetros adicionais do otimizador
x0 = 2*np.pi*np.random.random(2*N) # Conjunto inicial de parâmetros

result_PS = basinhopping(PS_fdf, x0, minimizer_kwargs=minimizer_kwargs, niter=n_iter) # Minimização

E_PS =  result_PS.fun # Energia minimizada
state_PS = result_PS.x # Parâmetros otimizados
```

```python
print(f'Energia do Ground-State: {E_PS}')
```

```python
nz = np.cos(state_PS[:N])**2 # Calculando o número de excitações
nx = (1 + np.sin(2*state_PS[:N])*np.cos(state_PS[N:]))/2 # Analogamente para sigmax
ny = (1 + np.sin(2*state_PS[:N])*np.sin(state_PS[N:]))/2 # Analogamente para sigmay
```

```python
# Plotando os números de excitação em cada ponto da rede
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, sharey=False, sharex=False, figsize=(15,5))

pc1 = ax1.scatter(vecs_rede[:,0], vecs_rede[:,1], c= nz, vmin=0., vmax=1.0, cmap=cm.bwr, s=100)

pc2 = ax2.scatter(vecs_rede[:,0], vecs_rede[:,1], c= nx, vmin=0., vmax=1.0, cmap=cm.summer, s=100)

pc3 = ax3.scatter(vecs_rede[:,0], vecs_rede[:,1], c= ny, vmin=0., vmax=1.0, cmap=cm.summer, s=100)

cb1 = fig.colorbar(pc1, ax=ax1, location='right')
cb2 = fig.colorbar(pc2, ax=ax2)
cb3 = fig.colorbar(pc3, ax=ax3)


ax1.axes.get_xaxis().set_visible(False)
ax1.axes.get_yaxis().set_visible(False)

ax2.axes.get_xaxis().set_visible(False)
ax2.axes.get_yaxis().set_visible(False)

ax3.axes.get_xaxis().set_visible(False)
ax3.axes.get_yaxis().set_visible(False)

ax1.set_title(r'$n_z$')
ax2.set_title(r'$n_x$')
ax3.set_title(r'$n_{y}$')

plt.tight_layout()
plt.show()
plt.close()
```

$\quad$ Vemos acima que agora parecemos ter encontrado o verdadeiro estado fundamental e, nele, os nossos átomos excitados e não-excitados se revezam, de forma a respeitar o mecanismo de bloqueio entre primeiros vizinhos devido à interação de Van der Waals.

# 3. Transições de Fase

$\quad$ Agora que sabemos como encontrar o estado fundamental do nosso sistema e calcular suas propriedades, podemos tentar buscar transições de fase ao variar a intensidade do detuning $\Delta$.

## 3.1. $\Omega = 0$

$\quad$ Tomemos primeiro o caso em que já estávamos trabalhando, onde $\Omega = 0$. Vamos definir uma lista de valores de $\Delta$ para analizar:

```python
Delta_list = np.arange(-1.0, 4., 0.3) # Definindo uma lista de valores de Delta para verificar como o sistema varia ao longo dela
```

$\quad$ Agora, computemos os valores da energia e do número de átomos excitados para cada valor na nossa `Delta_list`:

```python
E_list = [] # Incializando vetor para guardar as energias
sum_nz = [] # Inicializando vetor para guardar o número total de átomos excitados

for Delta in tqdm(Delta_list):
    
    n_iter = 500 # Número de iterações

    minimizer_kwargs = {"args":(Vij, Omega, Delta), "method":"L-BFGS-B", "jac":True} # Parâmetros adicionais
    x0 = 2*np.pi*np.random.random(2*N) # Parâmetros iniciais

    result_PS = basinhopping(PS_fdf, x0, minimizer_kwargs=minimizer_kwargs, niter=n_iter) # Minimização

    E_PS =  result_PS.fun # Energia minimizada 
    state = result_PS.x # Parâmetros otimizados
    
    E_list.append(E_PS) # Guardando as energias
    sum_nz.append(jnp.sum(np.cos(state[:N])**2)) # Guardando a soma de nz
    
```

```python
# Plotando as energias em função do detuning
fig = plt.figure()
plt.ylabel(r'$E_{GS}$', fontsize=15)
plt.xlabel(r'$\frac{\Delta}{C_6}$', fontsize = 15)
plt.plot(Delta_list, E_list, '-o')
```

```python
# Plotando os números de excitação em função do detuning
fig = plt.figure()
plt.ylabel(r'$\sum n_a$', fontsize = 15)
plt.xlabel(r'$\frac{\Delta}{C_6}$', fontsize = 15)
plt.plot(Delta_list, sum_nz, 'o')
```

$\quad$ Vemos, nas figuras acima, que a energia, como esperado, descresce conforme a razão $\frac{\Delta}{C_6}$ aumenta. O número total de átomos excitados, por outro lado, $\sum n_a$, se mantém sempre em platôs. Podemos verificar na variável `sum_nz` que esses platôs são sempre em números inteiros de excitações, como observamos antes:

```python
# Checando que o número de átomos excitado é sempre inteiro
sum_nz
```

$\quad$ Para verificar como o comportamento do sistema varia com o $\Delta$, podemos também calcular a susceptibilidade, definida como:

$$\chi = -\frac{d^2E}{d\Delta^2}, \tag{16}$$

```python
chi = -np.gradient(np.gradient(E_list, Delta_list), Delta_list) # Calculando a susceptibilidade
```

```python
# Plotando a susceptibilidade
plt.plot(Delta_list, chi)
plt.xlabel(r'$\frac{\Delta}{C_6}$', fontsize = 15)
plt.ylabel(r'$\chi$', fontsize = 15)
```

$\quad$ Vemos que a susceptibilidade é $0$ exceto por dois picos. Isso indica que as transições de fase entre fases com diferentes números de átomos excitados são de primeira ordem.

## 3.2. $\Omega \neq 0$

$\quad$ Agora, passemos ao caso com $\Omega \neq 0$. Para repetir a análise anterior para diferentes valores de $\Omega$, definamos a função `plots`, que dados valores de $\Omega$ e $C_6$, realiza a minimização para vários valores de $\Delta$ e plota os gráficos da energia, número de excitações e susceptibilidade em função de $\frac{\Delta}{C_6}$:

```python
# Definindo função para plotar as energias, números de excitações e susceptibilidades em função de Delta
def plots(Omega, C6):
    E_list = [] # Inicializando E
    sum_nz = [] # Inicializando n
    
    for Delta in tqdm(Delta_list):
        
        n_iter = 500 # Número de iterações
    
        minimizer_kwargs = {"args":(Vij, Omega, Delta), "method":"L-BFGS-B", "jac":True} # Parâmetros adicionais
        x0 = 2*np.pi*np.random.random(2*N) # Inicializando parâmetros aleatoriamente
    
        result_PS = basinhopping(PS_fdf, x0, minimizer_kwargs=minimizer_kwargs, niter=n_iter) # Minimizando
    
        E_PS =  result_PS.fun # Energia minimizada
        state = result_PS.x # Parâmetros otimizados
        
        E_list.append(E_PS) # Guardando a energia
        sum_nz.append(jnp.sum(np.cos(state[:N])**2)) # Guardando o número de átomos excitados
        

    # Plotando os gráficos de E e n
    fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, sharey=False, sharex=False, figsize=(15, 5))

    fig.suptitle(r'$\frac{\Omega}{C_6}$ =' + str(Omega), fontsize = 20)
    
    ax1.plot(Delta_list, E_list, '-o', markersize=5)
    ax1.set_ylabel(r'$E_{GS}$', fontsize = 15)
    ax1.set_xlabel(r'$\frac{\Delta}{C_6}$', fontsize = 15)
    
    ax2.plot(Delta_list, sum_nz, 'o',markersize=5)
    ax2.set_ylabel(r'$\sum n_a$', fontsize = 15)
    ax2.set_xlabel(r'$\frac{\Delta}{C_6}$', fontsize = 15)

    # Calculando e plotando a susceptibilidade chi
    chi = -np.gradient(np.gradient(E_list, Delta_list), Delta_list)
    
    ax3.set_xlabel(r'$\frac{\Delta}{C_6}$', fontsize = 15)
    ax3.set_ylabel(r'$\chi$', fontsize = 15)
    ax3.plot(Delta_list, chi)

    plt.tight_layout()
    plt.show()
    plt.close()
```

$\quad$ Agora, utilizemos esta função para realizar a mesma análise da seção 3.1. para os valores de $\Omega = 0.1$, $0.5$ e $1.0$, com $C_6 = 1$ fixo:

```python
Omega = 0.1 # Definindo Omega
```

```python
plots(Omega, C6) # Plotando para Omega = 0.1
```

```python
Omega = 0.5 # Definindo Omega
```

```python
plots(Omega, C6) # Plotando para Omega = 0.5
```

```python
Omega = 1.0 # Definindo Omega
```

```python
plots(Omega, C6) # Plotando para Omega = 1.0
```

$\quad$ Analisando os gráficos acima, vemos alguns comportamentos extremamente curiosos. Primeiro, o número de excitações total do sistema deixa de ser sempre inteiro e passa a tomar valores contínuos, fracionários conforme $\Delta$ varia. Isso se deve ao fato de que, ao tomar um valor de $\Omega \neq 0$, adicionamos algo análogo a um campo transversal ao sistema, introduzindo assim flutuações quânticas que quebram a simetria de número de excitações.

$\quad$ Observando também os gráficos de $\chi$, vemos que a susceptibilidade se afasta cada vez mais de seu comportamento inicial, não sendo mais nula fora dos picos, que já estavam presentes para $\Omega = 0$. Isso nos indica que as transições de fase que antes eram de primeira ordem agora são de segunda ordem, o que faz sentido, uma vez que agora o número de átomos excitados pode tomar, em cada sítio da rede, qualquer valor entre $0$ e $1$.

# 4. Diagrama de Fases

$\quad$ Tendo entendido o comportamento do sistema e aprendido a identificar suas transições de fase para um valor fixo de $C_6$, o próximo passo é construir e interpretar os diagramas de fase para vários valores dos parâmetros. Para isso, fixemos $\Omega = 1$ e definamos a função `E_ps_phases` para
calcular a energia e os parâmetros do estado fundamental para um par de valores de $R_b$ e $\Delta$. Nessa seção, utilizaremos a equação (7) para calcular o valor de $C_6$ a partir de $R_b$.

```python
# Definindo uma função para calcular as energias minimizadas e os parâmetros otimizados mais facilmente
def E_ps_phases(Rb, Delta):

    C6 = Omega*(Rb**6) # Calculando C6 a partir de Omega e do raio de bloqueio Rb
    Vij = jnp.array(V_int(C6, vecs_rede)) # Calculando as interações

    minimizer_kwargs = {"args":(Vij, Omega, Delta), "method":"L-BFGS-B", "jac":True} # Parâmetros adicionais

    x0 = 2*np.pi*np.random.random(2*N) # Parâmetros inicializados aleatoriamente

    result_mf = basinhopping(PS_fdf, x0, minimizer_kwargs=minimizer_kwargs, niter=500) # Resultados da minimização
    
    return (result_mf.fun, result_mf.x)
```

$\quad$ Agora, vamos definir os valores de $R_b$ e $\Delta$ que constituirão os pontos dos nossos diagramas de fases:

```python
Rb_list = np.arange(0.8,2.8,0.12) # Valores de Rb para percorrer
Delta_list = np.arange(-1.0, 4.0, 0.3) # Valores de Delta para percorrer
```

$\quad$ Por fim, vamos encontrar as propriedades do ground-state para todos os pares de pontos $(\Delta, R_b)$ nas nossas listas (é importante manter em mente que a célula abaixo pode demorar um bom tempo para rodar devido à grande quantidade de operações realizadas nela):

```python
E_mf = np.zeros((len(Delta_list), len(Rb_list))) # Matriz para guardar as energias em função de Delta e Rb
state_mf = np.zeros((len(Delta_list),len(Rb_list), 2*N)) # Matriz para guardar os parâmetros otimizados em função de Delta e Rb

Omega = 1.0 # Definindo o valor de Omega

for i, Delta in tqdm(enumerate(Delta_list)):

    for j, Rb in enumerate(Rb_list):
        E_ps, state_ps = E_ps_phases(Rb, Delta) # Energia e parâmetros otimizados
        E_mf[i][j] = E_ps # Guardando a energia
        state_mf[i][j] = state_ps # Guardando os parâmetros
```

$\quad$ Agora que nossos cálculos estão completos, resta apenas visualizar diferentes propriedades do nosso sistema com diagramas de fases. Faremos isso para quatro quantidades: a energia, a densidade de excitações $\rho_z = \frac{\sum n_a}{N}$, a susceptibilidade $\chi$ e a densidade análoga $\rho_x = \frac{\sum n_x}{N}$:

```python
# Plotando um diagrama de fases da energia em função de Rb e Delta
fig = plt.figure(figsize=(5,2.5), dpi=160)

pc = plt.pcolormesh(Delta_list, Rb_list, np.transpose(E_mf), shading='gouraud', vmin=np.min(E_mf), vmax=np.max(E_mf), cmap='jet')

plt.xlabel(r'$\frac{\Delta}{\Omega}$', fontsize=12)
plt.ylabel(r'$\frac{R_b}{a}$', fontsize=12)
fig.colorbar(pc)
plt.title(r'Energia do Ground-State, $\Omega = 1.0$', fontsize=12)

plt.show()
plt.clf()
plt.close()
```

```python
# Calculando as susceptibilidades em cada ponto do diagrama de fases
chi_mf = np.zeros((len(Rb_list), len(Delta_list)))
for i in range(len(Rb_list)):
    chi_mf[i] = -np.gradient(np.gradient(E_mf[:,i], Delta_list), Delta_list)
```

```python
# Plotando um diagrama de fases com as susceptibilidades
fig, ax1 = plt.subplots(figsize=(5.,2.5), dpi=160)

pc = ax1.pcolormesh(Delta_list, Rb_list, chi_mf, shading='gouraud', cmap='turbo')

ax1.set_xlabel(r'$\frac{\Delta}{\Omega}$', fontsize=12)
ax1.set_ylabel(r'$\frac{R_b}{a}$', fontsize=12)
cbar_ax = fig.add_axes([0.92, 0.15, 0.02, 0.7])
fig.colorbar(pc, cax=cbar_ax)

plt.suptitle(r' $\chi$, $\Omega = 1.0$', fontsize=12)
fig.subplots_adjust(right=0.9)


plt.show()
plt.clf()
plt.close()
```

```python
# Calculando os números de excitação
nz_mf = np.zeros((len(Delta_list), len(Rb_list)))

for i in range(len(Delta_list)):
    for j in range(len(Rb_list)):
        nz_mf[i][j] = np.sum(np.cos(state_mf[i][j][:N])**2)/N
```

```python
# Plotando um diagrama de fases com os números de excitação
fig, ax1 = plt.subplots(figsize=(5,2.5), dpi=160)

pc = ax1.pcolormesh(Delta_list, Rb_list, np.transpose(nz_mf), shading='gouraud', vmin=0, vmax=1, cmap='jet')

ax1.set_xlabel(r'$\frac{\Delta}{\Omega}$', fontsize=12)
ax1.set_ylabel(r'$\frac{R_b}{a}$', fontsize=12)
cbar_ax = fig.add_axes([0.92, 0.15, 0.02, 0.7])
fig.colorbar(pc, cax=cbar_ax)

fig.subplots_adjust(right=0.9)
plt.suptitle(r'$\rho_{n_z}, \Omega = 1.0$', fontsize=12)

plt.show()
plt.clf()
plt.close()
```

```python
# Calculando os nx para cada ponto do diagrama de fases
nx2_mf = np.zeros((len(Delta_list), len(Rb_list)))

for i in range(len(Delta_list)):
    for j in range(len(Rb_list)):
        nx2_mf[i][j] = np.sum((1 + np.sin(2*state_mf[i][j][:N])*np.cos(state_mf[i][j][N:]))/2)/N
```

```python
# Plotando um diagrama de fases com os nx
fig, ax1 = plt.subplots(figsize=(5,2.5), dpi=160)

pc = ax1.pcolormesh(Delta_list, Rb_list, np.transpose(nx2_mf), shading='gouraud', vmin=0, vmax=1, cmap='jet')

ax1.set_xlabel(r'$\frac{\Delta}{\Omega}$', fontsize=12)
ax1.set_ylabel(r'$\frac{R_b}{a}$', fontsize=12)
cbar_ax = fig.add_axes([0.92, 0.15, 0.02, 0.7])
fig.colorbar(pc, cax=cbar_ax)

fig.subplots_adjust(right=0.9)
plt.suptitle(r'$\rho_{n_x}, \Omega = 1.0$', fontsize=12)

plt.show()
plt.clf()
plt.close()
```

$\quad$ Observando os diagramas acima, notamos uma clara semelhança entre os diagramas que mostram $\chi$ e $\rho_x$. Ambos parecem ter transições bem marcadas entre regiões com propriedades qualitativamente diferentes. O fato de que ambos demonstram comportamentos análogos é um bom indício de que identificamos fases diferentes, que podem ser classificadas por analogia com sistemas de spin como fases ferromagnéticas, antiferromagnéticas e desordenadas conforme os comportamentos de $\rho_z$ e $\rho_x$ , que são análogos a magnetizações devido às suas relações com as matrizes de Pauli $\sigma^z$ e $\sigma^x$ (tradicionalmente interpretadas como operadores de spin em mecânica quântica).

$\quad$ A analogia acima é o que fundamenta o uso de redes de átomos de Rydberg como simuladores quânticos para sistemas de spin: a interação de Van der Waals, que faz com que átomos próximos tenham valores diferentes do número de excitação $n_a$ se comporta efetivamente como uma interação antiferromagnética entre spins, enquanto a interação mediada pelo detuning do laser tende a manter todos os átomos excitados, o que, por analogia, caracteriza um sistema ferromagnético.

$\quad$ Assim, alterar os parâmetros experimentais do laser altera a força dessas interações, permitindo a experimentalistas a realização de uma vasta gama de sistemas de spin em simuladores quânticos analógicos.
